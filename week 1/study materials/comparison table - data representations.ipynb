Here’s a compact comparison table you can study from:

| Representation      | Basic Structure                                            | Main Question It Answers                                 | What It Preserves                                         | What It Often Loses / Ignores                        | Common Data Mining Tasks                                                   | Example                                      |
| ------------------- | ---------------------------------------------------------- | -------------------------------------------------------- | --------------------------------------------------------- | ---------------------------------------------------- | -------------------------------------------------------------------------- | -------------------------------------------- |
| **Item Set**        | Unordered set of items                                     | **What is present?**                                     | Presence/absence of items, co-occurrence                  | Order, timing, often frequency                       | Frequent itemset mining, association rules, market basket analysis         | `{milk, bread, eggs}`                        |
| **Vector / Matrix** | Ordered list of feature values; many vectors form a matrix | **What features describe this object, and by how much?** | Numeric feature values, counts, weights, fixed dimensions | Usually order/sequence unless encoded separately     | Classification, clustering, similarity, regression, PCA                    | `<1, 0, 3, 2>` or a document-term matrix     |
| **Sequence**        | Ordered list of items/events                               | **In what order did things happen?**                     | Order, repetition, transitions                            | Exact numeric trends unless modeled separately       | Sequential pattern mining, next-step prediction, clickstream analysis      | `[search, view item, add to cart, purchase]` |
| **Time Series**     | Sequence of values indexed by time                         | **How does a value change over time?**                   | Temporal order, trends, seasonality, autocorrelation      | Rich relational structure, sometimes spatial context | Forecasting, anomaly detection, trend analysis, change-point detection     | Daily temperatures or hourly stock prices    |
| **Spatial**         | Objects/events with geographic coordinates or regions      | **Where is it?**                                         | Location, distance, geographic proximity                  | Time/order unless added separately                   | Spatial clustering, hotspot detection, nearest-neighbor analysis           | Crime incidents by latitude/longitude        |
| **Spatiotemporal**  | Data with both location and time                           | **Where and when did it happen?**                        | Movement, geographic change over time, trajectories       | Simplicity; can be complex and large-scale           | Trajectory mining, mobility analysis, outbreak tracking, traffic analysis  | GPS traces of delivery trucks                |
| **Graph / Network** | Nodes and edges representing entities and relationships    | **How are things connected?**                            | Connectivity, paths, neighborhoods, network structure     | Often exact geometric space or regular time patterns | Community detection, link prediction, centrality analysis, fraud detection | Social network of users and friendships      |
| **Stream**          | Continuous, potentially unbounded flow of incoming data    | **How do we process data as it arrives?**                | Real-time arrival, recent behavior, evolving patterns     | Full storage/history if memory is limited            | Online learning, drift detection, real-time anomaly detection              | Live sensor feed or transaction stream       |

## Quick way to remember it

| Representation  | Focus               |
| --------------- | ------------------- |
| Item Set        | Presence            |
| Vector / Matrix | Features and values |
| Sequence        | Order               |
| Time Series     | Change over time    |
| Spatial         | Location            |
| Spatiotemporal  | Location + time     |
| Graph / Network | Connections         |
| Stream          | Continuous arrival  |

## One more high-value distinction

| Pair                        | Key Difference                                                                                                   |
| --------------------------- | ---------------------------------------------------------------------------------------------------------------- |
| **Item Set vs Vector**      | Item sets usually show whether an item is present; vectors can also show how much of each feature is present     |
| **Sequence vs Time Series** | Sequences focus on ordered events/items; time series focus on numeric values over time                           |
| **Spatial vs Graph**        | Spatial data is about physical location; graph data is about relationships/connections                           |
| **Time Series vs Stream**   | Time series is a representation of temporal data; stream is a processing setting where data arrives continuously |

# Data Mining Functionalities

A **data mining functionality** is the kind of task you want to perform on data.
In other words, it answers:

> **What are we trying to discover or do with the data?**

Different functionalities serve different goals. Some describe data, some find hidden structure, and some make decisions or forecasts.

---

# 1. Online Analytical Processing (OLAP)

## What it is

**OLAP** is used to explore and summarize large amounts of data from different angles.

It is less about building a predictive model and more about **interactive analysis**.

## Main purpose

To answer questions like:

* What happened?
* How much?
* Where?
* When?
* In which category?

## Example

A company has sales data with dimensions such as:

* time,
* region,
* product,
* customer type.

With OLAP, an analyst can ask:

* Total sales by month
* Sales in California for Q1
* Sales by product category and region
* Compare this year vs last year

## Common OLAP operations

* **Roll-up**: summarize data at a higher level
  Example: daily sales → monthly sales
* **Drill-down**: go to more detail
  Example: yearly sales → quarterly → monthly
* **Slice**: select one dimension value
  Example: only data for 2025
* **Dice**: select a subcube of data
  Example: sales for California and Nevada in Q1 and Q2
* **Pivot**: rotate the view of the data
  Example: swap rows and columns in a report

## Strengths

* Fast summaries
* Easy business reporting
* Good for dashboards and decision support

## Limitation

OLAP tells you **what is in the data**, but not necessarily deeper hidden patterns or predictions.

# 2. Association

## What it is

**Association analysis** finds items or events that tend to occur together.

It is about discovering **co-occurrence patterns**.

## Main purpose

To answer:

* What tends to happen together?
* If X happens, what else is likely?

## Example

In grocery data:

* Customers who buy **bread** and **butter** often also buy **jam**

This can be written as a rule:

```text id="1ggv8c"
{bread, butter} → {jam}
```

## Common measures

* **Support**: how often the pattern appears in the dataset
* **Confidence**: how often the rule is true when the left side occurs
* **Lift**: how much stronger the relationship is than random chance

## Typical uses

* Market basket analysis
* Product bundling
* Recommender systems
* Web usage analysis

## Strengths

* Easy to interpret
* Useful for discovering hidden relationships

## Limitation

Association does **not** imply causation.
Just because two things occur together does not mean one causes the other.

---

# 3. Classification

## What it is

**Classification** assigns data objects to one of several **predefined categories**.

It is a **supervised learning** task, meaning the model learns from labeled examples.

## Main purpose

To answer:

* Which class does this object belong to?

## Example

A bank wants to classify loan applications as:

* **Approve**
* **Deny**

Or an email system classifies messages as:

* **Spam**
* **Not spam**

## How it works

You start with training data where the correct class labels are already known.
The model learns patterns from those labeled examples, then predicts the class of new cases.

## Common algorithms

* Decision trees
* Logistic regression
* Naive Bayes
* Support vector machines
* Neural networks
* k-nearest neighbors

## Typical uses

* Fraud detection
* Medical diagnosis
* Email filtering
* Customer churn detection

## Strengths

* Very useful when categories are known
* Can automate decision-making

## Limitation

Needs labeled training data, and performance depends on data quality and class definitions.

---

# 4. Prediction

## What it is

**Prediction** estimates an unknown or future value.

In many courses, classification is sometimes treated as a type of prediction, but often:

* **classification** predicts a category
* **prediction** predicts a numeric or continuous value

## Main purpose

To answer:

* What value should we expect?

## Example

Predict:

* house price,
* tomorrow’s temperature,
* customer lifetime value,
* future sales.

## Common algorithms

* Linear regression
* Regression trees
* Random forests
* Gradient boosting
* Neural networks
* Time series models

## Typical uses

* Sales forecasting
* Risk scoring
* Demand estimation
* Price prediction

## Strengths

* Helps planning and decision-making
* Can estimate future outcomes

## Limitation

Predictions are uncertain and can fail when the future differs from the past.

---

# 5. Clustering

## What it is

**Clustering** groups similar objects together **without predefined labels**.

It is an **unsupervised learning** task.

## Main purpose

To answer:

* Are there natural groups in the data?
* Which objects are similar to each other?

## Example

A company has customer data but no customer types defined yet.
Clustering may reveal groups such as:

* budget shoppers,
* loyal premium customers,
* infrequent discount buyers.

## Common algorithms

* k-means
* Hierarchical clustering
* DBSCAN
* Gaussian mixture models

## Typical uses

* Customer segmentation
* Document grouping
* Image grouping
* Social network community discovery

## Strengths

* Useful when labels are unavailable
* Can reveal hidden structure

## Limitation

Clusters may not always be meaningful, and different algorithms can produce different groupings.

---

# 6. Ranking

## What it is

**Ranking** orders objects from highest to lowest according to relevance, importance, preference, or likelihood.

## Main purpose

To answer:

* Which items should appear first?
* Which option is best?
* Which cases deserve priority?

## Example

A search engine ranks webpages by relevance to a query.
A hiring system ranks applicants by fit.
An online store ranks products for recommendation.

## Types of ranking goals

* Relevance ranking
* Priority ranking
* Risk ranking
* Recommendation ranking

## Example applications

* Search engines
* Recommendation systems
* Credit risk prioritization
* Triage systems

## Common methods

* Scoring models
* Learning-to-rank algorithms
* PageRank-style graph methods
* Pairwise ranking models

## Strengths

* Helpful when users need the “best few” items first
* Very useful in search and recommendation systems

## Limitation

Ranking quality depends on the scoring logic and relevance definition.

---

# 7. Outlier / Anomaly Detection

## What it is

This functionality identifies data objects that are **unusual**, **rare**, or **very different** from the rest.

## Main purpose

To answer:

* What does not fit the normal pattern?
* Which cases look suspicious or exceptional?

## Example

* A credit card purchase far outside a customer’s normal spending pattern
* A machine sensor reading that suddenly spikes
* A network login from an unusual country at an unusual time

## Types of anomalies

* **Point anomaly**: one unusual data point
* **Contextual anomaly**: unusual in a specific context
  Example: 90°F is normal in summer, strange in winter
* **Collective anomaly**: a group of values is abnormal together

## Common methods

* Statistical thresholds
* Distance-based methods
* Density-based methods
* Isolation Forest
* One-class SVM
* Autoencoders

## Typical uses

* Fraud detection
* Intrusion detection
* Equipment failure monitoring
* Quality control

## Strengths

* Crucial for risk and safety systems
* Can find rare but important cases

## Limitation

Not all outliers are meaningful, and some true anomalies are hard to distinguish from noise.

---

# How these functionalities differ

Here is the main comparison:

| Functionality         | Main Goal                          |   Supervised? | Example Output                 |
| --------------------- | ---------------------------------- | ------------: | ------------------------------ |
| **OLAP**              | Summarize and explore data         |            No | Sales by region and month      |
| **Association**       | Find co-occurring patterns         |            No | `{milk, bread} → {butter}`     |
| **Classification**    | Assign a category                  |           Yes | Spam / Not spam                |
| **Prediction**        | Estimate a numeric or future value |   Usually yes | House price = $450,000         |
| **Clustering**        | Find natural groups                |            No | 4 customer segments            |
| **Ranking**           | Order items by score or relevance  | Can be either | Top 10 most relevant pages     |
| **Outlier Detection** | Find unusual cases                 |    Usually no | Suspicious transaction flagged |

---

# A simple way to remember them

* **OLAP** → summarize
* **Association** → co-occur
* **Classification** → label
* **Prediction** → estimate
* **Clustering** → group
* **Ranking** → order
* **Outlier detection** → flag unusual cases

---

# One example dataset, many functionalities

Suppose you have retail store data.

## OLAP

“Show total sales by month, store, and product category.”

## Association

“Customers who buy cereal often also buy milk.”

## Classification

“Will this customer churn: yes or no?”

## Prediction

“How much will this customer spend next month?”

## Clustering

“What customer segments naturally exist?”

## Ranking

“Which products should be recommended first?”

## Outlier detection

“Which transactions look fraudulent?”

This is why data mining is broad: the **same dataset** can support many different functionalities depending on the question.

---

# Important distinctions

## Classification vs prediction

* **Classification** predicts a **category**
* **Prediction** often predicts a **number/value**

## Clustering vs classification

* **Clustering** finds groups without labels
* **Classification** assigns objects to known labels

## OLAP vs data mining

* **OLAP** is mainly for summarizing and exploring known dimensions
* **Data mining** often goes further to discover hidden patterns or build predictive models

## Ranking vs classification

* **Classification** assigns a class
* **Ranking** orders candidates by priority or relevance

## Outlier detection vs clustering

* **Clustering** looks for common group structure
* **Outlier detection** looks for rare exceptions

---

# Bottom line

These functionalities answer different kinds of questions:

* **OLAP**: What happened?
* **Association**: What occurs together?
* **Classification**: Which class is this?
* **Prediction**: What value will happen?
* **Clustering**: What groups exist?
* **Ranking**: What should come first?
* **Outlier detection**: What is unusual?

# Data Mining Functionalities Comparison Table


| Functionality                           | Core Purpose                                        | Typical Question                        | Supervised or Unsupervised                             | Typical Output                               | Example Use Case                               |
| --------------------------------------- | --------------------------------------------------- | --------------------------------------- | ------------------------------------------------------ | -------------------------------------------- | ---------------------------------------------- |
| **OLAP (Online Analytical Processing)** | Summarize and explore data across dimensions        | “What happened, where, and when?”       | Neither in the ML sense; mainly exploratory/analytical | Aggregated reports, cubes, pivoted summaries | Sales by region, quarter, and product category |
| **Association**                         | Find items/events that occur together               | “What tends to happen together?”        | Unsupervised                                           | Association rules such as `A → B`            | Customers who buy bread also buy butter        |
| **Classification**                      | Assign objects to predefined classes                | “Which category does this belong to?”   | Supervised                                             | Class label                                  | Spam vs not spam, approve vs deny loan         |
| **Prediction**                          | Estimate a future or unknown numeric value          | “What value should we expect?”          | Usually supervised                                     | Numeric value or forecast                    | Predict house price or next month’s sales      |
| **Clustering**                          | Group similar objects without labels                | “What natural groups exist?”            | Unsupervised                                           | Clusters or segments                         | Customer segmentation                          |
| **Ranking**                             | Order items by relevance, importance, or likelihood | “Which should come first?”              | Can be supervised or unsupervised                      | Ordered list / scores                        | Search results, product recommendations        |
| **Outlier / Anomaly Detection**         | Identify unusual or rare cases                      | “What does not fit the normal pattern?” | Usually unsupervised, sometimes supervised             | Outlier flag, anomaly score                  | Fraud detection, unusual sensor readings       |

## Another version focused on differences

| Functionality         | Focus                            | Learns From Labels? | Main Strength                                | Main Limitation                                          |
| --------------------- | -------------------------------- | ------------------: | -------------------------------------------- | -------------------------------------------------------- |
| **OLAP**              | Exploration and summarization    |                  No | Fast multidimensional analysis               | Does not usually discover deep hidden patterns by itself |
| **Association**       | Co-occurrence relationships      |                  No | Easy-to-interpret rules                      | Does not imply causation                                 |
| **Classification**    | Known categories                 |                 Yes | Strong for decision-making when labels exist | Needs labeled training data                              |
| **Prediction**        | Numeric estimation / forecasting |        Yes, usually | Useful for planning and forecasting          | Sensitive to changing future conditions                  |
| **Clustering**        | Hidden group structure           |                  No | Reveals natural segments                     | Results may depend heavily on method/settings            |
| **Ranking**           | Prioritization / ordering        |           Sometimes | Great when only top results matter           | Depends on how relevance is defined                      |
| **Outlier Detection** | Rare or abnormal cases           |          Usually no | Useful for risk, fraud, monitoring           | Can confuse noise with meaningful anomalies              |

## Fast distinction examples

| Compare                             | Key Difference                                                                                                |
| ----------------------------------- | ------------------------------------------------------------------------------------------------------------- |
| **Classification vs Prediction**    | Classification predicts a category; prediction usually predicts a numeric value                               |
| **Classification vs Clustering**    | Classification uses known labels; clustering finds groups without labels                                      |
| **OLAP vs Data Mining Models**      | OLAP explores and aggregates known dimensions; data mining models often discover patterns or make predictions |
| **Ranking vs Classification**       | Ranking orders many candidates; classification assigns a class                                                |
| **Clustering vs Outlier Detection** | Clustering looks for common group structure; outlier detection looks for rare exceptions                      |

Here’s a clean way to think about **data representations in data mining**:

**Data representation** is the way raw data is structured so we can analyze it, measure patterns, compare objects, and build models.
The representation you choose matters because it determines:

* what information is preserved,
* what kinds of patterns you can detect,
* what algorithms will work well,
* and what information may be lost.

A useful rule is:

> **The representation should match the structure of the problem.**

Below is a practical overview of each type.

---

# 1. Item Set

## What it is

An **item set** represents each object as an **unordered collection of items**.

It usually answers: **What is present?**

## Example

A grocery basket:

```text
Transaction 1 = {milk, bread, eggs}
Transaction 2 = {bread, butter}
Transaction 3 = {milk, eggs}
```

A document can also be represented as an item set of words:

```text
Doc = {data, mining, michigan, course}
```

## Key characteristics

* Unordered
* Usually ignores frequency unless extended
* Focuses on presence/absence of items

## What it is good for

* Market basket analysis
* Association rules
* Frequent pattern mining

## Common tasks

* Frequent itemset mining
* Association rule mining
* Co-occurrence analysis

## Strengths

* Simple
* Easy to interpret
* Good for “which items appear together?”

## Limitations

* Loses order
* Often loses repetition/count information
* Not ideal when timing or sequence matters

## Example insight

“If customers buy **bread** and **butter**, they often also buy **jam**.”

---

# 2. Vector / Matrix

## What it is

A **vector** represents one object as a list of values across dimensions.
A **matrix** stacks many vectors together.

It usually answers: **How much of each feature is present?**

## Example

Suppose vocabulary is:

```text
<michigan, data, science, mining, course>
```

Then a document might be:

```text
D1 = <1, 1, 1, 0, 0>
D2 = <0, 1, 0, 1, 1>
```

Each position is a feature.

## Matrix form

If you have many documents, you get a document-term matrix:

| Document | michigan | data | science | mining | course |
| -------- | -------- | ---- | ------- | ------ | ------ |
| D1       | 1        | 1    | 1       | 0      | 0      |
| D2       | 0        | 1    | 0       | 1      | 1      |

## Key characteristics

* Ordered dimensions
* Can store binary, counts, continuous values
* Supports algebra and geometry

## What it is good for

* Classification
* Clustering
* Similarity calculations
* Dimensionality reduction
* Regression

## Common tasks

* k-means clustering
* Logistic regression
* SVM
* PCA
* Cosine similarity

## Strengths

* Flexible
* Works with many machine learning methods
* Can capture counts, magnitudes, and numeric features

## Limitations

* Can become very sparse
* High-dimensional data may be expensive
* May ignore order unless explicitly encoded

## Example insight

Two documents are similar because their vectors point in similar directions.

---

# 3. Sequence

## What it is

A **sequence** is an **ordered list of events or items**.

It answers: **In what order did things happen?**

## Example

Customer purchase history:

```text
<Customer A> = [browse laptop, compare prices, add to cart, purchase]
```

DNA:

```text
[A, C, G, T, A, A, C]
```

Website clicks:

```text
[home, search, product page, checkout]
```

## Key characteristics

* Order matters
* Repetition matters
* Gaps may matter too

## What it is good for

* Sequential pattern mining
* Clickstream analysis
* Bioinformatics
* User behavior modeling

## Common tasks

* Sequence mining
* Next-event prediction
* Sequential classification

## Strengths

* Preserves temporal or logical order
* Useful for behavior and process data

## Limitations

* More complex than itemsets
* Similarity is harder to define
* Long sequences can be noisy

## Example insight

Users who do `search → reviews → cart` are more likely to purchase.

---

# 4. Time Series

## What it is

A **time series** is a sequence of values measured over time, usually at regular or meaningful intervals.

It answers: **How does a quantity change over time?**

## Example

```text
Daily temperature = [72, 74, 70, 68, 71, 75]
Hourly heart rate = [80, 85, 90, 88, 84]
Stock price over time
```

## Key characteristics

* Time index is essential
* Trend, seasonality, cycles, and noise matter
* Nearby observations are often correlated

## What it is good for

* Forecasting
* Trend detection
* Anomaly detection
* Change-point analysis

## Common tasks

* Time series forecasting
* Smoothing
* Decomposition
* Pattern matching

## Strengths

* Captures change over time
* Useful for monitoring and prediction

## Limitations

* Time dependence violates assumptions of many standard models
* Missing values and irregular sampling can be difficult
* Noise can hide patterns

## Example insight

Electricity usage spikes every weekday morning and evening.

---

# 5. Spatial

## What it is

**Spatial data** represents objects with a location in physical space.

It answers: **Where is it?**

## Example

* Store locations on a map
* Crime incidents by latitude/longitude
* Land parcels
* Hospital locations

A point might be:

```text
(34.093, -118.127)
```

## Key characteristics

* Location matters
* Distance matters
* Nearby things may be related

## What it is good for

* Geographic analysis
* Hotspot detection
* Nearest-neighbor queries
* Regional clustering

## Common tasks

* Spatial clustering
* Spatial interpolation
* Density estimation
* Geographic pattern mining

## Strengths

* Captures geographic relationships
* Supports location-based insights

## Limitations

* Requires careful handling of distance and scale
* Boundaries and coordinate systems matter
* Spatial autocorrelation complicates analysis

## Example insight

Disease cases are clustered in one part of the city.

---

# 6. Spatiotemporal

## What it is

**Spatiotemporal data** combines **location + time**.

It answers: **Where and when did something happen?**

## Example

* GPS traces of vehicles
* Spread of wildfire over time
* Migration of animals
* Traffic congestion across a city throughout the day

An observation might be:

```text
(time = 8:15 AM, lat = 34.093, lon = -118.127, speed = 42 mph)
```

## Key characteristics

* Both time and location matter
* Patterns may evolve over time and space together

## What it is good for

* Mobility analysis
* Route mining
* Outbreak tracking
* Environmental monitoring

## Common tasks

* Trajectory mining
* Movement prediction
* Event detection in space and time

## Strengths

* Rich and realistic for many real-world systems
* Can reveal dynamic geographic patterns

## Limitations

* Complex
* Large data volumes
* Harder to visualize and model

## Example insight

Traffic builds first in one corridor, then spreads outward during rush hour.

---

# 7. Graph / Network

## What it is

A **graph** represents entities as **nodes** and relationships as **edges**.

It answers: **How are things connected?**

## Example

Social network:

```text
Nodes: people
Edges: friendships
```

Web graph:

```text
Nodes: webpages
Edges: hyperlinks
```

Protein interaction network:

```text
Nodes: proteins
Edges: interactions
```

## Key characteristics

* Relationship structure is central
* Can be directed or undirected
* Can have weights, labels, or attributes

## What it is good for

* Social network analysis
* Recommendation systems
* Fraud detection
* Community detection
* Link prediction

## Common tasks

* Centrality analysis
* Community detection
* Shortest paths
* PageRank
* Graph embeddings

## Strengths

* Captures relational structure directly
* Powerful for connected systems

## Limitations

* Can be computationally expensive
* Harder to visualize at scale
* Not well handled by many standard tabular methods

## Example insight

A few highly central accounts spread information much faster than others.

---

# 8. Stream

## What it is

A **data stream** is data arriving continuously, often rapidly, and potentially without end.

It answers: **How do we analyze data as it arrives?**

## Example

* Sensor feeds
* Social media posts
* Online transactions
* Real-time website clicks
* IoT devices

## Key characteristics

* Continuous arrival
* Potentially unbounded
* Must often be processed in real time
* Limited memory/storage constraints

## What it is good for

* Real-time anomaly detection
* Online learning
* Monitoring systems
* Fraud detection
* Event detection

## Common tasks

* Sliding-window analysis
* Streaming classification
* Drift detection
* Approximate counting

## Strengths

* Supports real-time decisions
* Useful for live systems

## Limitations

* Cannot usually store everything
* Algorithms must be fast and incremental
* Concept drift can change patterns over time

## Example insight

A credit card system flags an unusual purchase pattern seconds after it occurs.

---

# How these representations differ

Here is the main distinction among them:

| Representation | Main Question                                      |
| -------------- | -------------------------------------------------- |
| Item set       | What items are present?                            |
| Vector/Matrix  | What features and values describe the object?      |
| Sequence       | In what order did events occur?                    |
| Time series    | How does a value change over time?                 |
| Spatial        | Where is the object/event?                         |
| Spatiotemporal | Where and when did it happen?                      |
| Graph/Network  | How are entities connected?                        |
| Stream         | How do we process data continuously as it arrives? |

---

# How representation affects data mining

This is the most important takeaway.

## Same data, different representation, different insights

Take shopping data:

* As an **item set**: find products often bought together
* As a **sequence**: find the order in which products are bought
* As a **time series**: study sales trends over weeks
* As a **graph**: connect customers to products and analyze communities
* As a **stream**: detect suspicious purchases in real time

So the “best” representation depends on the question.

---

# Quick intuition for when to use each

Use **item sets** when you care about **co-occurrence**.
Use **vectors/matrices** when you care about **features and numeric modeling**.
Use **sequences** when **order** matters.
Use **time series** when **change over time** matters.
Use **spatial** when **location** matters.
Use **spatiotemporal** when **location and time together** matter.
Use **graphs** when **relationships and connections** matter.
Use **streams** when data must be handled **continuously and immediately**.

# One simple example across multiple representations

Suppose you are studying public transit riders.

## Item set

Each rider trip:

```text
{bus, downtown, transfer}
```

Use: common trip characteristics

## Vector

Each rider:

```text
<age, fare, trip_count, avg_distance>
```

Use: clustering rider types

## Sequence

Each rider’s trip history:

```text
[home stop → bus line 4 → station → line B]
```

Use: route behavior

## Time series

Daily ridership counts:

```text
[1200, 1305, 1180, 1450, ...]
```

Use: forecasting demand

## Spatial

Stops as coordinates on a map
Use: underserved areas

## Spatiotemporal

Bus positions over time
Use: delay patterns by place and time

## Graph

Stations as nodes, routes as edges
Use: important hubs and bottlenecks

## Stream

Live arrival data
Use: real-time service alerts

# Common exam-ready distinctions

These are worth remembering:

## Item set vs vector

* **Item set** usually captures only presence/absence.
* **Vector** can capture counts, weights, and numeric values.

## Sequence vs time series

* **Sequence** focuses on ordered events/items.
* **Time series** focuses on values indexed by time.

## Spatial vs graph

* **Spatial** is about physical location.
* **Graph** is about connectivity, which may or may not be geographic.

## Time series vs stream

* **Time series** is a data form.
* **Stream** is a processing setting where data arrives continuously.

---

# Bottom line

In data mining, **representation is not just formatting**. It determines what patterns are visible and what methods make sense.

A strong way to remember it is:

* **Item set** → presence
* **Vector/matrix** → features
* **Sequence** → order
* **Time series** → change over time
* **Spatial** → location
* **Spatiotemporal** → location + time
* **Graph/network** → connections
* **Stream** → continuous arrival

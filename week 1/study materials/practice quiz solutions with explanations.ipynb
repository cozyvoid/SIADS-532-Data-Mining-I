# Practice Quiz Solutions with Explanations

## 1. Which representation is most natural for modeling prerequisite relationships among courses?

**Answer: C. Graph/Network**

**Explanation:**
A prerequisite relationship connects one course to another in a dependency structure. That is exactly what a graph represents well. Each course can be a node, and a directed edge can show that one course must be completed before another. Itemsets and vectors do not naturally represent dependency relationships.

---

## 2. Which data objects can reasonably be represented as a graph/network?

**Answer: A, D**

**Explanation:**

* **A. Courses and their prerequisites**: This is a classic graph because courses depend on one another.
* **D. Course co-enrollment relationships**: This can also be modeled as a graph, where two courses are connected if students often take them together.

Why not the others?

* **B. Students and the terms in which they enrolled** is better handled as sequences or relational records.
* **C. Paths as sets of required courses** is more naturally an itemset unless you also want to encode prerequisite dependencies.

---

## 3. If Alice represents each student by the set of courses they have taken, ignoring term order and grades, what representation is this?

**Answer: B. Itemset**

**Explanation:**
An itemset is simply a collection of items with no ordering. If the only information kept is which courses a student took, then the student is represented as a set of courses. No timing, sequence, or repeated structure is preserved.

---

## 4. If Alice wants to preserve the order of terms in which a student took courses, which representation is best?

**Answer: C. Sequence**

**Explanation:**
A sequence preserves order. That makes it ideal when Alice wants to know not just what courses a student took, but the progression over time. Since course histories unfold term by term, sequence data is the best match.

---

## 5. Why might representing a student’s course history as an itemset lose important information?

**Answer:** Because itemsets ignore order and timing.

**Explanation:**
An itemset only tells us which courses were taken. It does not tell us:

* when courses were taken,
* in what order they were taken,
* whether the student followed a sensible prerequisite progression.

Two students could have the exact same itemset but very different academic trajectories.

---

## 6. What are advantages of representing students as binary course vectors?

**Answer: A, C, D**

**Explanation:**

* **A. Easy input for many machine learning models**: Many algorithms work naturally with vectors.
* **C. Supports similarity comparisons between students**: Students with similar vectors took similar courses.
* **D. Easy to include many courses in a uniform format**: Every student is represented in the same feature space.

Why not **B**?
Vectors do not preserve prerequisite order unless extra features are added.

---

## 7. Which representation would best capture student progress over time across terms?

**Answer: B. Sequence**

**Explanation:**
The key idea is that student course-taking behavior has an order. A sequence directly captures this. A time series is more appropriate if Alice were measuring a numeric variable over time, like GPA by term or number of credits per semester. But course histories themselves are sequences of course events.

---

## 8. Give one way to represent a path.

**Answer:** A path could be represented as an itemset of required courses, plus a prerequisite graph over those courses.

**Explanation:**
A path has two important aspects:

* the set of courses that belong to it,
* the prerequisite structure among those courses.

Using both representations preserves both membership and dependency.

---

## 9. Alice wants to discover which courses are frequently taken together before students choose a path. Which functionality is most appropriate?

**Answer: B. Association pattern mining**

**Explanation:**
Association mining finds items that tend to co-occur. In this case, the “items” are courses, and Alice wants to know which ones are commonly taken together. This is very similar to market basket analysis.

---

## 10. Alice wants to predict whether a student is likely to complete a given path within X years. Which functionality best fits?

**Answer: A. Classification**

**Explanation:**
This is a yes/no prediction problem:

* complete within X years,
* not complete within X years.

That makes it a classification task.

---

## 11. Alice wants to group students into types such as “early deciders,” “explorers,” and “late path committers.” Which functionality is most appropriate?

**Answer: B. Clustering**

**Explanation:**
Clustering groups similar data objects without predefined labels. Alice is not starting with labeled student types; she wants to discover natural groups based on enrollment behavior.

---

## 12. Which of the following are examples of prediction tasks in this scenario?

**Answer: A, C**

**Explanation:**

* **A. Predicting whether a student will graduate within X years** is prediction.
* **C. Predicting a student’s likely path based on their first few terms** is also prediction.

Why not the others?

* **B. Discovering common course combinations** is descriptive pattern mining.
* **D. Detecting unusual enrollment trajectories** is outlier detection.

---

## 13. Alice wants to find common orderings of courses that successful students follow. Which functionality fits best?

**Answer: B. Sequential pattern mining**

**Explanation:**
Sequential pattern mining is specifically for finding frequent ordered patterns. Since Alice wants orderings like A → B → C, this is the right tool.

---

## 14. Alice wants to identify students whose enrollment histories are very unusual compared with others. Which functionality should she use?

**Answer: A. Outlier detection**

**Explanation:**
Outlier detection identifies data objects that do not fit normal patterns. In this case, unusual enrollment paths, strange timing, or very atypical course combinations could signal outliers.

---

## 15. Why is a graph representation better than a simple itemset for representing course prerequisites?

**Answer:** Because a graph preserves dependency structure.

**Explanation:**
An itemset can tell Alice which courses are involved, but not how they depend on one another. A graph adds structure: it shows which courses must come before others. That dependency information is central to prerequisites.

## 16. Suppose Alice wants to recommend the next course a student should take based on what similar students did in the past. This is most closely related to:

**Answer: A. Recommendation / similarity-based mining**

**Explanation:**
This is similar to recommendation systems. Alice would compare a student to past students with similar histories and use those patterns to suggest a likely next course.

## 17. Which features might be useful in a model predicting path completion?

**Answer: A, B, C**

**Explanation:**

* **A. Number of prerequisite courses already completed**: Strong signal of progress.
* **B. Terms elapsed since entering the program**: Important for the X-years deadline.
* **C. Whether the student has ever changed intended path**: Could affect delay or uncertainty.

## 18. If Alice uses student-course enrollment records and term-by-term order to model academic behavior, she is primarily working with:

**Answer: B. Sequence data**

**Explanation:**
The defining property here is order through time. That makes it sequence data. Even though the course system may also involve graphs, the enrollment history itself is best described as sequential.

---

## 19. What is one limitation of representing course histories only as vectors?

**Answer:** Vectors ignore order and timing.

**Explanation:**
A vector can tell Alice which courses a student has taken, but not whether the student took them early, late, or in a sensible order. That can hide important academic patterns.

---

## 20. If Alice wants to predict a student’s time to graduation, where the outcome is numeric, which functionality is most appropriate?

**Answer: B. Regression**

**Explanation:**
When the target is a number, such as 3.5 years or 12 terms, the task is regression rather than classification.

---

## 21. Alice wants to design a new path by looking for groups of courses that successful graduates commonly complete together. What representation and functionality might she use?

**Answer:** Itemsets plus frequent pattern mining.

**Explanation:**
If Alice wants to find groups of courses that commonly appear together among successful students, each student can be represented as an itemset of completed courses. Then frequent pattern mining can reveal common course groupings.

## 22. Alice wants to verify that a proposed path is feasible within X years. What kinds of data and analysis would help?

**Answer:** Prerequisite graph, course schedules, and historical student progress data.

**Explanation:**
Alice would need:

* the **prerequisite structure** to know valid course order,
* the **term availability of courses** to know when each can actually be taken,
* **historical enrollment patterns** to see how long similar students took.

Together, these help determine whether a path is realistically completable within X years.

## 23. What best describes the difference between representing a student’s history as an itemset versus a sequence?

**Answer: B. Sequences preserve order, itemsets do not**

**Explanation:**
This is the core distinction:

* **itemset** = what was taken,
* **sequence** = what was taken and in what order.

---

## 24. Alice may want to combine multiple representations because:

**Answer: A, B, D**

**Explanation:**

* **A. Different tasks require different information**: Advising and path design are not the same task.
* **B. No single representation captures all useful aspects**: One format usually loses something.
* **D. Enrollment history has both structural and temporal aspects**: Structure comes from prerequisites; timing comes from terms.

Graphs are powerful, but they do not automatically replace itemsets, vectors, or sequences.

## 25. Give one example of a classification label Alice could use.

**Explanation:**
A classification label is a discrete category. In this scenario, many advising questions can be turned into yes/no outcomes, which makes them classification targets.

## 26. If two students took the same set of courses but in different orders, which representation would distinguish them?

**Answer: C. Sequence**

**Explanation:**
An itemset or vector would treat both students as identical because both only record course membership. A sequence would distinguish them because it preserves order.

## 27. If Alice wants to model the prerequisite system where Course A must come before Course B, the edge in the graph should most naturally be:

**Answer: B. Directed from A to B**

**Explanation:**
If A is required before B, then the dependency flows from A to B. This direction encodes that B depends on A.

## 28. Which questions are best answered by sequential pattern mining?

**Answer: B, C**

**Explanation:**
Sequential pattern mining looks for frequent ordered behavior.

* **B. What course orders are common among students who finish a path?** Yes.
* **C. What courses tend to be taken before students declare a path?** Yes.

Why not **A**?
That is about co-occurrence, not order, so association mining is better.

Why not **D**?
That is outlier detection.

## 29. Why might clustering students be useful for advising?

**Answer:** Because it can reveal groups of students with similar academic trajectories.

**Explanation:**
For example, clustering might identify:

* students who decide quickly,
* students who explore widely before choosing,
* students who struggle with prerequisite bottlenecks.

These groups can help advisors provide more tailored guidance instead of giving every student the same advice.

---

## 30. Why is this a good example of needing both descriptive and predictive mining?

**Answer:** Because Alice needs both discovery and forecasting.

**Explanation:**

* **Descriptive mining** helps her understand the data: common course combinations, common sequences, prerequisite bottlenecks, student clusters.
* **Predictive mining** helps her forecast outcomes: whether a student will finish on time, which path they are likely to choose, what next course is best.

The scenario naturally involves both understanding patterns and making decisions based on future outcomes.

---

# Big Takeaway

This scenario is useful because it shows that **data representation drives what questions can be asked well**.

For example:

* **Itemsets** help answer “what courses go together?”
* **Sequences** help answer “in what order do students progress?”
* **Graphs** help answer “what dependency structure exists?”
* **Vectors** help support machine learning and similarity analysis

And once the data is represented appropriately, Alice can apply the right mining functionality:

* **association mining** for co-occurring courses,
* **sequential pattern mining** for common progressions,
* **clustering** for student groups,
* **classification/regression** for prediction,
* **outlier detection** for unusual cases.

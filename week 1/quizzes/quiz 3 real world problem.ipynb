## Note: Use the following scenario for all questions in this quiz.

### *Alice’s school is offering a new undergraduate program and its curriculum is organized into a few "paths".  A path (e.g., an analyst's path, a user experience path) contains the same number of courses (assuming that they are equal in credit hours).  A student needs to finish at least one path to graduate within X years, and beyond that path they can take any other courses. The school allows students to take some courses first before deciding on which path to take - that can be several terms into the program.* 

### *In each path, certain courses must be taken before certain others (known as prerequisites), but otherwise the order of courses does not matter. There can be multiple prerequisites for one course, and the same course can be prerequisite of multiple courses.*

### *Alice is charged to design new paths for the program and also advise students for course selection. Let’s assume that all the courses already exist (no new courses will be added) and Alice has access to the complete history of course enrollment records of the school (who has taken which course in which term). Alice wants to accomplish the task through data mining.*  

### *The first thing Alice needs to do is to decide on the representations of the data objects: courses, students, course enrollment history, grades, and paths.  Then she can formulate her tasks as data mining functionalities.*

## In the following five questions, you will help Alice represent five relevant data objects in her context as five different data representations.

### Q1. What type of data representation should Alice use to represent a course?

- a sequence
- a vector
- a time series 
**- an item**
- a matrix 

For **Q1**, the correct answer is **an item**.

**Why:**
A single **course** is one basic entity in Alice’s context. It is not itself an ordered list, a numeric feature array, or a time-based record. Instead, it is one atomic object that can later appear inside larger representations, such as:

* a student’s completed courses as an **itemset**
* a term-by-term enrollment history as a **sequence**
* prerequisite relations among courses as a **graph**

Here, a course is a categorical (instead of numerical) variable.

So the course itself is best represented as **an item**.

### 2. What type of data representation should Alice use to represent final grades (in percentage) of one course?

- a sequence
- **a vector**
- a time series
- an item
- a matrix 

The final grade of one student in one course is a numerical attribute. So a vector is the natural representation for the final grades of a course, with each student enrolled in the course as a dimension.

For **Q2**, the correct answer is **a vector**.

**Why:**
The **final grades (in percentage) of one course** would be a collection of numeric values, one per student who took that course. A vector is the natural way to represent a list of numbers for one object or variable.

For example, if five students took the course, the grade representation could look like:

`[92, 85, 78, 88, 95]`

That is a **vector**.

Why not the others?

* **sequence**: would emphasize ordered events, which is not the main idea here
* **time series**: would require values indexed over time
* **item**: too simple for a whole set of grades
* **matrix**: would be more appropriate if Alice were representing grades for **multiple courses across multiple students**

So the best choice is **a vector**.

### 3. What type of data representation should Alice use to represent course grades of all students who graduated from the same path?

- a sequence
- a vector
- a time series
- an item
**- a matrix**

Since the final grades of one course can be represented as a vector, then the final grades of multiple courses (in a path) can be represented as a matrix (with columns being students and rows being courses). You can also represent every course as a dimension, and then every student becomes a row in the matrix.

Yes — **a matrix** is the correct answer for **Q3**.

**Why:**
Now Alice is not looking at grades for just **one course** anymore. She wants the **course grades of all students who graduated from the same path**. That means there are:

* **multiple students**
* **multiple courses**
* a grade value for each student-course combination

That is naturally represented as a **matrix**, where:

* **rows** = students
* **columns** = courses
* **entries** = grade percentages

Example:

\begin{bmatrix}
90 & 85 & 88 \
78 & 92 & 84 \
95 & 87 & 91
\end{bmatrix}

This could represent 3 students across 3 courses in the same path.

Why not the others?

* **vector**: only works well for one list of numbers, like grades for one course
* **sequence**: would emphasize order, which is not the main focus here
* **time series**: requires time-indexed numeric data
* **item**: too simple for a whole set of student-course grades

So for **grades across many students and many courses**, the best representation is **a matrix**.


### 4. What type of data representation should Alice use to represent a student’s course enrollment history?

**- a sequence**
- a vector
- a time series
- an item
- a matrix 

A student's course enrollment history consists of courses (items) in order (terms). A sequence is a natural representation for this type of data. A data stream can also be a valid choice, although it is not among the options here.

Yes — **a sequence** is the correct answer for **Q4**.

**Why:**
A student’s **course enrollment history** is inherently **ordered over time**. The important information is not just which courses the student took, but **the order in which they took them**, usually by term.

Example:

`(Math 101, CS 101) → (Stats 200) → (ML 300, UX 210)`

That is a **sequence** because it captures progression across semesters or terms.

Why not the others?

* **vector**: would show features or course indicators, but not order
* **time series**: usually means numeric values observed over time, like GPA each term
* **item**: too small; a whole history is not a single item
* **matrix**: could store enrollments, but does not naturally emphasize ordered progression as well as a sequence

So the best representation for a student’s course enrollment history is **a sequence**.

### 5. What type of data representation should Alice use to represent the enrollment number of the same course over the years?

- a sequence
- a vector
**- a time series**
- an item
- a matrix 

Because the enrollment number of a course is a numerical attribute that changes over time (terms), it is natural to represent it as a time series.

Yes — **a time series** is the correct answer for **Q5**.

**Why:**
Alice is representing the **enrollment number of the same course over the years**. That means she has:

* one variable: enrollment count
* measured repeatedly
* across time

That is exactly what a **time series** is.

Example:

* 2021: 85 students
* 2022: 91 students
* 2023: 76 students
* 2024: 103 students

This is a sequence of **numeric values indexed by time**, which makes it a **time series**.

Why not the others?

* **sequence**: captures ordered events or items, but not specifically numeric measurements over time
* **vector**: could store the values, but it would miss the explicit time-indexed meaning
* **item**: too simple
* **matrix**: would be more appropriate if Alice were comparing many courses across many years at once

So the correct answer is **a time series**.

Across the five questions, the mapping is:

1. **course** → **an item**
2. **final grades of one course** → **a vector**
3. **course grades of all students in the same path** → **a matrix**
4. **a student’s course enrollment history** → **a sequence**
5. **enrollment number of the same course over years** → **a time series**

### 6. Now it's time to represent a "path" in the program. A path is intuitively a structure of multiple courses. But what's the suitable representation here?

### True or False: Alice may represent a "path" as a vector.

**- False**


A vector does not preserve the order information in a path.

For **Q6**, the correct answer is **False**.

**Why:**
A **path** is not just a collection of courses. In this scenario, it also includes **prerequisite structure** among those courses. A plain **vector** can list features or indicate whether courses belong to the path, but it does **not naturally capture the dependency relationships** between courses.

For example, a vector could show:

* Course A = 1
* Course B = 1
* Course C = 1

But it would not show that:

* A must come before B
* A must come before C
* B and C may both be prerequisites for D

That structure is much better represented as a **graph/network**, specifically a **directed graph**, where:

* **nodes** = courses
* **directed edges** = prerequisite relationships

So the statement **“Alice may represent a path as a vector”** is **False** because a vector loses the key structural information that defines the path.

### 7. True or False: Alice may represent a "path" as one sequence.

### (Hint: The prerequisite structure in the given scenario might make a difference here)

**- False**

Order does matter between courses, so it is easy to think about using sequence representation. However, because there may be multiple prerequisites of a course and one course may be the prerequisite of multiple courses, one sequence alone can not represent this information.

For **Q7**, the correct answer is **False**.

**Why:**
A **path** should **not** be represented as **one sequence** because the courses in a path do **not** necessarily have one fixed order.

In the scenario:

* some courses must come before others because of **prerequisites**
* but **otherwise the order does not matter**
* a course can have **multiple prerequisites**
* one course can be a prerequisite for **multiple other courses**

That means the path has a **partial order**, not a single linear order.

A sequence would force Alice to choose one exact ordering, such as:

`A → B → C → D`

But the real path may allow several valid orderings, for example:

* `A → B → C`
* `A → C → B`

if B and C only both depend on A and not on each other.

So representing a path as **one sequence** would incorrectly imply that there is only one valid order. The better representation is a **directed graph/network**, which can capture all prerequisite constraints without forcing one linear sequence.

So the statement is **False**.

### 8. True or False: Alice may represent a "path" as a network.

**- True**

Order does matter between courses, so we need to represent the relation between courses. Because there may be multiple prerequisites of a course and one course may be the prerequisite of multiple courses, a network would be a suitable representation of the relations. Also, a sequence is a special case of a network, so whatever a sequence can represent, a network can too. 

For **Q8**, the correct answer is **True**.

**Why:**
A **path** can be represented as a **network** because the key structure in a path is the **prerequisite relationship among courses**.

In this representation:

* **nodes** = courses
* **edges** = prerequisite relationships

More specifically, it is best thought of as a **directed network/graph**, since direction matters:

* if Course A must be taken before Course B, then the edge goes from **A → B**

This works well because a path does **not** always have one fixed sequence. Instead, it often has:

* multiple valid ways to progress
* branching prerequisite relationships
* courses with multiple prerequisites
* courses that unlock several later courses

A network captures all of that without forcing one single order.

So the statement **“Alice may represent a path as a network”** is **True**.

### 9. True or False: Alice may represent a "path" as a time series.

**- False**

Students aren’t required to take a course at a certain time, as long as they take the courses in the right order. That’s the difference between time series and sequence/network representation of data.

For **Q9**, the correct answer is **False**.

**Why:**
A **time series** is used for **numeric values measured over time**, such as:

* enrollment count by year
* average GPA by semester
* graduation rate over time

But a **path** is not primarily a time-indexed numeric signal. It is a **set of courses with prerequisite relationships**. The important structure is **dependency**, not a sequence of measurements over time.

A time series would not naturally show:

* which courses belong to the path
* which courses must come before others
* branching prerequisite structure
* multiple valid ways to complete the path

That is why a **network/graph** is appropriate for a path, while a **time series** is not.

So the statement **“Alice may represent a path as a time series”** is **False**.

### 10. True or False: Alice may represent a "path" as multiple sequences.

**- True**

Order does matter between courses, so it is easy to think about sequence representation. However, because there may be multiple prerequisites of a course and one course may be the prerequisite of multiple courses, one sequence alone can hardly represent this information. But we can use multiple sequences to represent this data (a path = all qualifying sequences of courses)

For **Q10**, the correct answer is **True**.

**Why:**
A path cannot usually be represented well by **one single sequence**, because the prerequisite structure allows **multiple valid course orders**. But it **can** be represented as **multiple sequences**, where each sequence is one valid linear ordering of the courses that satisfies the prerequisite constraints.

For example, suppose:

* A must come before B
* A must come before C
* B and C can be taken in either order

Then the path could be represented by multiple valid sequences such as:

* `A → B → C`
* `A → C → B`

So:

* **one sequence** is generally not enough
* **multiple sequences** may represent the different valid ways to complete the path

That said, this is still not the most compact or natural representation overall. A **network/graph** is usually better because it captures the prerequisite structure directly without listing all valid orderings.

So the statement **“Alice may represent a path as multiple sequences”** is **True**.

### 11. Bob is a current student in the program, but he has not determined his path yet. Alice tries to recommend the courses for Bob to take next semester.

### Correct or Incorrect: Alice may do this through finding the courses that are compliant (in the correct order) with Bob's current course enrollments on as many paths as possible. By doing this he has the largest flexibility to choose potential paths.

**- Correct**

By doing this, Bob has the flexibility to explore as many paths as possible before he makes his decision. This is a sequence prediction task.

**Correct** is the right answer for **Q11**.

**Why:**
Bob has **not chosen a path yet**, so Alice should avoid recommending courses that lock him too early into only one option if that is unnecessary. A smart strategy is to recommend courses that:

* satisfy prerequisite order
* are useful for **multiple possible paths**
* keep as many future options open as possible

That gives Bob the **largest flexibility** while still making progress.

In data mining terms, Alice could use historical enrollment and prerequisite information to identify courses that are:

* valid next steps given Bob’s current history, and
* common or compatible across several paths

That makes the recommendation strategy described in the question **correct**.

### 12. Correct or Incorrect: Alice may do this through finding courses that are commonly taken by former students with course enrollment history that are “similar” to Bob’s.

**- Correct**

By doing this, the courses Bob takes are most likely to be useful for the potential paths he would pursue (because students who have taken the same courses as Bob tended to proceed to these courses). This is also a sequence prediction task.

**Correct** is the right answer for **Q12**.

**Why:**
This is a classic **similarity-based recommendation** idea. Alice can look at former students whose course enrollment histories are similar to Bob’s and then identify which courses those students commonly took next.

That makes sense because:

* students with similar histories are likely at similar stages
* their next-course choices may be good candidates for Bob
* historical enrollment data can be mined to support this recommendation

This is essentially using past student trajectories as guidance for recommending future courses.

One caveat: in practice, Alice should still check that the recommended courses:

* satisfy Bob’s prerequisites
* fit semester availability
* do not reduce his flexibility unnecessarily

But as a general data mining strategy, the statement is **correct**.

### 13. Correct or Incorrect: Alice may do this through choosing the courses that have the largest enrollments.

**- Incorrect**

This approach does not consider Bob’s course enrollment history at all, and the course is likely to be not relevant to the potential paths that Bob would pursue.

**Incorrect** is the right answer for **Q13**.

**Why:**
Choosing courses only because they have the **largest enrollments** is not a good recommendation strategy for Bob.

High enrollment may simply mean a course is:

* popular,
* required for many students,
* easy to access,

But that does **not** mean it is the right next course for Bob specifically.

A good recommendation should depend on things like:

* Bob’s current enrollment history
* prerequisite satisfaction
* compatibility with multiple possible paths
* what similar students took next

A large-enrollment course could still be:

* irrelevant to Bob’s likely paths
* unavailable because he lacks prerequisites

So enrollment size alone is not enough. That is why the statement is **incorrect**.

### 14. The next thing Alice does is to recommend paths to students who have enrolled in the program for a few terms, based on their interest, course enrollment history, and the feasibility of finishing in time.

### Is the following procedure reasonable for the task?

### Measure the similarity between a student’s current course enrollment history and each (possible sequence of each) path, and then return the most similar path(s) to students.

**- Yes, this is reasonable.** (I disagree.)

Per instructer:

By doing this, you are casting the problem as a ranking task and you are solving it through computing the similarities between sequences. A path more similar to the student's current course sequence is more likely to be relevant to their interest, and also more likely to be accomplishable in time.

Why I disagree:

**No, this is not reasonable.**

**Why:**
The problem is the phrase **“each possible sequence of each path.”** A path is not naturally one fixed sequence, and trying to expand a path into **all possible valid course sequences** is usually a bad idea.

Here is why this is not a good procedure:

* A **path is really a prerequisite network/partial order**, not a single ordered list.
* One path may have **many valid sequences** that satisfy its prerequisites.
* Enumerating **all possible sequences** can become very large and inefficient.
* Similarity to one of those sequences may not reflect the true fit of the **path as a whole**.
* The task also depends on **interest** and **feasibility of finishing on time**, not just sequence similarity.

A better approach would be to compare the student’s history to the **path structure itself**, such as:

* courses already completed that belong to the path
* prerequisite requirements already satisfied
* remaining workload
* whether the path can still be completed within the allowed time
* alignment with the student’s interests

So the procedure in the question is **not reasonable** as stated.

## Is the following procedure reasonable for the task?

### 16. Treat each path as a category and classify a student’s current course enrollment history into these categories based on student's interest, the set of courses they have taken, and the orders between certain courses (e.g., infovis I and infovis II) they have taken.

**- Yes, this is not reasonable.**

By doing this, you are casting the problem as a classification task, and you are solving it through extracting the patterns from the itemset or sequence data.

**Yes, this is reasonable.**

**Why:**
This procedure fits the task much better than the previous one because it treats **each path as a category** and uses meaningful features about the student to predict which path is the best match.

It uses:

* the student’s **interest**
* the **set of courses** already taken
* the **order** of relevant courses in the enrollment history

Those are all useful signals for recommending a path.

Why this makes sense:

* A path recommendation can naturally be framed as a **classification problem**.
* Each possible path is a **class label**.
* A student’s background and history become the **input features**.
* The model can learn from past students which patterns tend to align with each path.

One important note:
To fully match the task, Alice should also include features related to **feasibility of finishing in time**, such as how many required courses remain or how many terms are left. But the overall procedure is still a **reasonable** approach.

So the correct answer is **Yes, this is reasonable.**

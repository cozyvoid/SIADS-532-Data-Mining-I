# Module 1: Quiz 1, Part 1 - Basic Concepts

### 1. Both time series and data streams are sensitive to timestamps that are associated with the data objects.

**- True**

The difference is that in a (single) time series, every timestamp is associated with a single numerical value, and in data stream every timestamp is associated with a complex data object.

Both **time series** and **data streams** depend on the **time/timestamp associated with observations**, because the ordering and timing of data points affect how the data is interpreted and analyzed.

- **Time series**: timestamps are central because the goal is usually to study values over time.
- **Data streams**: timestamps matter because data arrives continuously, and recency/order often affects processing and decisions.

### 2. The difference between classification and clustering is that clustering uses similarity between data objects but classification doesn’t.

**- False**

Both classification and clustering can be built upon similarity between data objects (think about nearest neighbors). The difference between the two data mining functions lies in whether the categories are predefined and whether there are labeled examples available to train the classifier.

The key difference is:

- **Classification** assigns data objects to **predefined labels/classes**.
- **Clustering** groups data objects into **natural groupings without predefined labels**.

Clustering does rely heavily on **similarity** between objects, but classification can also use similarity in some methods, such as **k-nearest neighbors**. So similarity is **not** the defining difference.

### 3. Patterns and similarities are two basic outputs of data mining that many advanced data mining functionalities can be built upon.

**- True**

Patterns and similarities are two basic functionalities, although they are not the *only* two outputs. Many of the more complex data mining outputs/functionalities can be built upon them.

**Patterns** and **similarities** are foundational outputs in data mining.

- **Patterns** include things like association rules, frequent itemsets, and trends.
- **Similarity** helps compare objects and is used in tasks like clustering, nearest-neighbor methods, and anomaly detection.

Many more advanced data mining tasks are built from these two ideas.

### 4. Text data can only be represented as itemsets or sequences, as words are natural items.

**- False**

Text data can also be represented as vectors, networks, or other structures. Words can be represented as dimensions of a vector or nodes of a network, for example.

Text data can be represented in many ways, not just as **itemsets** or **sequences**.

Examples:

- **Itemsets**: treating words as unordered terms
- **Sequences**: preserving word order
- **Vectors/matrices**: bag-of-words, TF-IDF, document-term matrices
- **Embeddings**: numerical vector representations of words, sentences, or documents
- **Graphs**: relationships between words or documents

So words may be natural items, but text is not limited to only those two representations.

### 5. Unlike database systems, machine learning has nothing to do with data mining.

**- False**

Databases and machine learning are both related to data mining, depending on which view you take. Machine learning provides methodology for building data mining functionalities.  Data mining provides data representation, feature extraction, and other fundamental tools for machine learning.

Machine learning is closely related to **data mining**.

- **Data mining** is the broader process of discovering useful patterns, trends, and knowledge from data.
- **Machine learning** provides many of the methods used in data mining, especially for tasks like **classification, prediction, clustering, and anomaly detection**.

So unlike database systems, machine learning is not separate from data mining; it is one of the major technical foundations used within it.

# Module 1 Quiz 2, Part 2: Data Representations

### 1. Different data representations are sometimes transformable from one another. For example, itemsets and vectors can sometimes be used as alternatives to represent the same type of data.

### We can transform a collection of itemsets into the same number of vectors such that: 1) the vectors share the same dimensions; 2) each unique item becomes a dimension of the vectors; and 3) for each vector transformed from an itemset, the value under one dimension is either 1 (the item appears in the set) or 0 (the item doesn’t appear in the set). 

### Let’s consider a dataset that contains three text documents, each of which has only a few words. The three documents have been represented as itemsets as below:

### - D1 = {michigan, data, science},

### - D2 = {data, mining, course},

### - D3 = {course, michigan}.

### Just to represent the 3 documents above as vectors (or a matrix), how many dimensions do we need to create (assuming that there isn't any other word or document in the dataset)?

You need one dimension for each unique word across the three documents:

1. michigan
2. data
3. science
4. mining
5. course

So the vector space has **5 dimensions**.

### 2. Suppose we define the dimensions as <michigan, data, science, mining, course>. 

### Then after transformation, we have D1 = <1, 1, 1, 0, 0> and D2 = <0, 1, 0, 1, 1,>. 

### What representation do we have for D3?

**- <1, 0, 0, 0, 1>**

Using the unique words across all three documents, the full vocabulary is:

{michigan, data, science, mining, course}

- Now convert each document into a binary vector in that order:

- D1 = {michigan, data, science}
→ (1, 1, 1, 0, 0)
- D2 = {data, mining, course}
→ (0, 1, 0, 1, 1)
- D3 = {course, michigan}
→ (1, 0, 0, 0, 1)

- Because D3 = {course, michigan}, and the dimension order is:

⟨michigan, data, science, mining, course⟩

So:

- michigan = 1
- data = 0
- science = 0
- mining = 0
- course = 1

- Therefore, **D3 = ⟨1, 0, 0, 0, 1⟩**

### 3. In reality, itemsets and vectors have their pros and cons to represent this type of data. Please consider the following statements and select all that are true:

Both statements are true.

- **True**: An itemset only records whether a word is present, not how many times it appears, so repeated words are lost.
- **True**: A vector over a large vocabulary is often very sparse, meaning it contains many zeros, which can waste storage.

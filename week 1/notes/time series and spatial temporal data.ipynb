### Time-Series and Spatial/Temporal Data

### Time Series Data

- **Data Object**: growth chart, stock prices over time, battery life over time



- **Attribute**: the measurement of a (*numerical*) property observed at a *given time point*

### The Time Series Representation

- A list of timestamped measurements:

$
X = {(x_1, t_1), (x_2, t_2),..., (x_n)}
$

- $x_i$ is the (*numerical*) measurement of a property of $X$ observed at time stamp $t_i$
- Alternative representation: $ x = f(t) $

### Examples of Time Series

- Stock Market (SP 500)



### Spatial/Spatiotemporal Data

- **Data Object**: GPS trajectory of a vehicle, spread of a disease, a heat map
- **Attribute**: measurement of a (numerical) property at a given location is **spatial data**. If measurement also includes a given time point, it is **spatiotemporal data**.



Location can be:
- 1D = just name of location
- 2D = longitude and latitude
- 3D = longtitude, latitude, altitude

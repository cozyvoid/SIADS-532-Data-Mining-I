### 1. What are the differences between using patterns to classify an unknown data object versus using similarity?

Using patterns to classify an unknown data object means assigning it to a class based on whether it matches learned rules, features, or characteristic structures associated with that class. The focus is on identifying distinguishing properties that are strongly associated with known categories. In other words, pattern-based classification asks: Does this object contain the important signals that define a class?

Using similarity to classify an unknown object means comparing it directly to known examples and assigning it to the class of the most similar objects. The focus is less on explicit rules and more on closeness or resemblance. Similarity-based classification asks: Which known objects does this new object look most like?

The main difference is that pattern-based classification relies on discovered class-defining structures, while similarity-based classification relies on distance or resemblance to labeled examples.

Example:
Suppose we want to classify an email as spam or not spam.

A pattern-based approach may look for signals such as “contains many promotional phrases,” “has suspicious links,” or “uses certain word combinations common in spam.”
A similarity-based approach may compare the email to previously labeled emails and classify it according to the most similar ones.

So, pattern methods emphasize rules and discriminative traits, while similarity methods emphasize comparison and neighborhood relationships.

### 2. What are the differences between using patterns to identify clusters of data objects versus using similarity?

Using patterns to identify clusters means grouping objects based on shared structures, recurring characteristics, or common feature combinations. The goal is to find groups defined by meaningful common patterns. This approach asks: What common properties make these objects belong together?

Using similarity to identify clusters means grouping objects that are close to one another according to some distance or similarity measure. The goal is to form groups where members are more alike to each other than to objects outside the group. This approach asks: Which objects are most alike overall?

The main difference is that pattern-based clustering looks for shared internal structure, while similarity-based clustering looks for closeness in representation space.

Example:
Imagine customer shopping data.

A pattern-based clustering approach might group customers because they share recurring purchase behaviors, such as frequently buying organic food and wellness products together.
A similarity-based clustering approach might group customers based on how close their purchase vectors are across many categories.

So, with clustering:

Patterns explain why objects belong together in terms of common structure.
Similarity groups objects based on how close they are, even if the exact shared pattern is not explicitly stated.

### A concise summary

- Patterns for classification: use rules or characteristic features to assign a label.
- Similarity for classification: assign a label based on closeness to known labeled examples.
- Patterns for clustering: find groups defined by shared recurring structure.
- Similarity for clustering: find groups of objects that are near each other in feature space.

A useful way to think about it is that patterns are more explanation-oriented, while similarity is more distance-oriented.

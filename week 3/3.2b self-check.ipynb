### 1. How do Manhattan distance (review) and Euclidean distance (review) differ from another?

**Manhattan distance** measures the distance between two vectors by adding the absolute differences across each dimension. It is like moving along a grid, only horizontally and vertically.

**Euclidean distance** measures the straight-line distance between two vectors in space. It is based on the square root of the sum of squared differences.

So:

- **Manhattan distance** = grid-like path distance
- **Euclidean distance** = straight-line distance

Manhattan distance tends to treat each dimension more independently, while Euclidean distance gives more weight to larger differences because the differences are squared.

### 2. What are the overall strengths of using Manhattan or Euclidean distances? How do these compare to using a dot product? (review)

The strength of **Manhattan** and **Euclidean** distances is that they measure **how far apart** two data points are. This makes them useful for clustering, nearest-neighbor methods, and cases where actual closeness in space matters.

- **Manhattan distance** is often useful when movement or differences happen dimension by dimension and can be more robust to large single-dimension differences.
- **Euclidean distance** is useful when straight-line geometric closeness is the most natural measure.

Compared with these, the **dot product** does not directly measure distance. Instead, it measures **alignment** between two vectors based on their values. Larger values in the same dimensions increase the dot product.

So:

- **Manhattan / Euclidean** tell you how **different or far apart** two vectors are
- **Dot product** tells you how strongly two vectors **point together or align**

In general, distances are better when you care about **closeness**, while dot product is better when you care about **similarity in direction and shared magnitude**.

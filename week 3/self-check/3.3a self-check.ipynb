### 1. How do eigenvectors help measure patterns in matrices? What features of matrices can lead to different patterns? (review)

Eigenvectors help identify the main **directions of variation or structure** in a matrix. They show the directions along which the data changes most consistently. Their corresponding eigenvalues indicate **how strong** those patterns are.

In data mining, this means eigenvectors help reveal important patterns such as:

- directions of greatest spread in the data
- groups of variables that change together
- hidden structure in a matrix

Features of matrices that can lead to different patterns include:

- **correlation** between columns or rows
- differences in **scale or variance**
- repeated or redundant information
- whether the data is dense or sparse
- the overall arrangement of values across rows and columns

So, eigenvectors help summarize the most important structural patterns in a matrix.

### 2. How do correlation and orthogonality relate to the need of eigendecomposition? (review)

If variables in a matrix are highly **correlated**, they contain overlapping information. This makes the data harder to interpret because multiple dimensions may be expressing similar patterns.

**Orthogonality** means directions are at right angles to each other, so they are independent in a geometric sense. In matrix analysis, orthogonal directions reduce redundancy and separate distinct sources of variation.

Eigendecomposition helps because it transforms the matrix into a set of directions where the main patterns are separated out. These directions are often orthogonal, which makes the structure easier to analyze. In other words, eigendecomposition helps turn correlated structure into a cleaner set of independent pattern directions.

### 3. What are principal components and how do they relate to orthogonality? (example or definition)

**Principal components** are the new axes created in PCA that capture the major variation in the data.

- The **first principal component** captures the greatest amount of variance.
- The **second principal component** captures the next greatest amount of variance, subject to being orthogonal to the first.
- Additional components continue this pattern.

Their relation to orthogonality is that **each principal component is orthogonal to the others**. This means each one captures a new, non-overlapping source of variation.

So, principal components are orthogonal directions that summarize the most important variation in the data.

### 4. PCA (Principal Component Analysis) will transform an inputted matrix into a new matrix where all coordinates are orthogonal to each other. How does using eigenvectors and eigendecomposition help accomplish that? (review)

PCA usually starts from the covariance matrix of the data. Then it performs **eigendecomposition** on that matrix.

This gives:

- **eigenvectors**, which become the new axes or directions
- **eigenvalues**, which tell how much variance each direction explains

Because the covariance matrix is symmetric, its eigenvectors are orthogonal. PCA uses these orthogonal eigenvectors as a new coordinate system. The original data is projected onto these new axes, producing a transformed matrix where the coordinates are uncorrelated and aligned with the main patterns in the data.

So, eigendecomposition provides the exact orthogonal directions PCA needs to rotate the data into a cleaner representation.

### 5. What is SVD (Singular Value Decomposition) able to accomplish that PCA cannot? (review)

SVD is more general than PCA.

PCA is typically applied to a **covariance matrix** and is mainly used to find principal components for variance-based dimensionality reduction.

SVD can be applied directly to **any matrix**, even if it is:

- rectangular
- not symmetric
- not a covariance matrix

This means SVD can handle a wider range of data analysis tasks. For example, it is especially useful for:

- decomposing non-square data matrices
- low-rank approximation
- noise reduction
- latent semantic analysis in text mining
- recommendation systems

PCA is often derived using SVD, but SVD can do more because it works on the original matrix itself, not just the covariance matrix.

So, the main advantage is that **SVD works for more general matrices and broader applications**, while PCA is more specifically focused on finding orthogonal directions of maximum variance.

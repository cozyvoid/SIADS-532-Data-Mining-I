### 1. How is calculating similarity between vectors different than calculating similarity between itemsets? (review)

Similarity between **itemsets** is based on shared items, usually focusing on **whether an item is present or not**. It does not usually account for quantity or weight. A common measure is **Jaccard similarity**, which compares the size of the intersection to the size of the union.

Similarity between **vectors** uses the actual **numeric values** in each dimension. This means it can capture not just whether two objects share features, but also **how much** they share and in what proportions. Common vector similarity measures include **dot product** and **cosine similarity**.

So, itemset similarity is mainly about **overlap in membership**, while vector similarity is about **numerical relationship across dimensions**.

### 2. How does dot product differ from using set intersections? (review) 

A **set intersection** only checks which items appear in both sets. It treats items as either present or absent and ignores counts or magnitudes.

A **dot product** multiplies corresponding values in two vectors and then adds the results. Because of this, it reflects both:

- whether the same dimensions are nonzero
- how large the values are in those shared dimensions

For example:

- Set intersection asks: **What do these two objects both contain?**
- Dot product asks: **How strongly do these two objects align across their shared dimensions?**

So, set intersection is a **binary overlap measure**, while dot product is a **numeric similarity measure** that includes magnitude

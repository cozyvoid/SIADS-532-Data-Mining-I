### 1. What are the limitations of itemsets that sometimes make vectors a better form of data representation? For which cases are itemsets better, and for which cases are vectors better? (review)

Itemsets are limited because they usually only represent whether an item is present or absent. They do not naturally capture **frequency, magnitude, weight, or intensity**. For example, if a word appears 10 times in a document versus 1 time, an itemset may treat both cases the same. Itemsets also do not preserve numeric relationships between features.

Vectors are better when you need to represent **how much** of something there is, not just whether it exists. A vector can store counts, measurements, ratings, probabilities, or other numeric values for each dimension.

Itemsets are better for:

- simple presence/absence data
- market basket data
- association rule mining
- cases where only membership matters

Vectors are better for:

- text data with word counts or weights
- numeric feature data
- similarity calculations based on magnitude
- machine learning tasks that use continuous or count-based variables

So, itemsets are better for **qualitative membership-style data**, while vectors are better for **quantitative, feature-based data**.

### 2. What does the magnitude and direction of a vector in coordinate space represent? (review)

The **magnitude** of a vector represents its **length or overall size**. It tells you how large or strong the vector is.

The **direction** of a vector represents **where it points in the coordinate space**. It shows how the values are distributed across dimensions.

In data terms:

- magnitude can reflect the **overall amount** of something
- direction can reflect the **pattern or composition** of the data across features

For example, two document vectors might point in a similar direction if they contain similar proportions of words, even if one document is much longer.

### 1. What are the limitations of using dot product, Manhattan, and Euclidean distances? (review)

Each method has limitations:

- **Dot product** is affected by **magnitude**. Two vectors can have a large dot product just because they are long, even if their patterns are not especially similar.
- **Manhattan distance** and **Euclidean distance** measure how far apart two vectors are, but they are also sensitive to **scale**. Large values can dominate the result.
- **Euclidean distance** especially can be strongly influenced by large differences in one dimension because it squares the differences.
- All three methods can be less useful when you care more about the **relative pattern** of values than the absolute size.

So, these methods may give misleading results when vectors differ a lot in length or scale.

### 2. What is cosine? What problem does it solve regarding previously discussed methods? (review)

**Cosine** usually refers to **cosine similarity**, which measures the cosine of the angle between two vectors.

Instead of focusing on absolute magnitude or raw distance, cosine focuses on **direction**. This helps solve the problem that dot product, Manhattan distance, and Euclidean distance can be overly affected by vector length or scale.

For example, if two document vectors have the same overall pattern of word usage but one document is much longer, cosine similarity can still show that they are very similar.

### 3. What are the overall advantages of using cosine compared to other methods? (review)

The main advantages of cosine similarity are:

- it focuses on **orientation/pattern** rather than raw size
- it reduces the effect of **magnitude**
- it works well for **high-dimensional sparse data**, such as text
- it is useful when the important question is whether two vectors are similar in **composition**

Compared to dot product, cosine avoids favoring longer vectors just because they have larger values. Compared to Manhattan and Euclidean distance, cosine is better when absolute distance matters less than the overall pattern across features.

So, cosine is especially useful when you want to compare **shape or proportion** rather than **amount**.

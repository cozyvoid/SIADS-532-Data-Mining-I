# Vector Similarity Functions and Dot Product

### What is "similarity"?

- Similarity is a measyre of how much two data objects are alike
- Distance measures the opposite: how much they are dissimilar

### Similarity/Distance of Vectors





- If more non-empty dimensions in common -> more similar
- If values are closer on each shared dimension -> larger the similarity
- Handle real values
- Normalization is often needed

### Dot Product



Sum of dimension-wise similarity
- Zero if either $x_i$ or $y_i$ is zero
- Positive if $x_i$ and $y_i$ have the same sign
- Negative if $x_i$ and $y_i$ have different signs

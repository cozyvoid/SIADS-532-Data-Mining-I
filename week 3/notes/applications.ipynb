# Applications

### Vector similarity for complex data mining tasks



- Classification
- Clustering
- Ranking
- Recommendation
- Visualization

- Many similarity and distance functions available
- Selection depends on data and task

### Applications of vector similarity

- **Similarity of text**
    - Information retrieval (search engines)
- **Similarity of ratings**
    - Recommender systems, information filtering
- **Machine Learning**
    - Minimize distance between the vector of predicted values and the vector of ground truth labels
    - Manhattan distance and Euclidean distance are related to the L1 and L2 norms

- How to measure the similarity and distance between vectors
- How to calculate dot product, cosine similarity, Manhattan distance, Euclidean distance, and Pearson coefficient
- Select and use vector similarity and distances in real applications

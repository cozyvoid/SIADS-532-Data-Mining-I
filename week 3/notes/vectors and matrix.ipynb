# Vectors and Matrix

### Vector representation of data



- Denoted as $X$ or $x$ if no ambiguity
- $p-$ dimensional vector, for $p$ attributes
- Dimensionality of $X$ is $p$
- $x_i$ can be any real numbers
- Called binary vector if all $x_j \epsilon {0, 1} $

### Matrix data



A list of vectors with the same dimensionality
- $n$ denotes the number of vectors in a matrix
- $p$ denotes the number of dimensions
- $X$ is a $n \times p$ ($n$ by $p$) matrix; $n$ rowa, $p$ columns

### Row vector and column vector



- A row vector represents a data object
- A column vector represents an attribute
- A matrix is a collection of row and column vectors

### Transpose of matrix





- $p \times x$ matrix
- Row vectors become column vectors
- Column vectors become row vectors

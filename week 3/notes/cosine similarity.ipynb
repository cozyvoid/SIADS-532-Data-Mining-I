# Cosine Similarity

### Problems with Manhattan and Euclidean Distance

Vectors are not normalized



- Are vectors with larger values likely to yield a higher or lower distance?
- What about dot product?

### Example

- Three documents: 
    - D1 contains "data mining"
    - D2 contains "data mining data mining"
    - D3 contains "data"

- Is D1 closer to D2 or D3?
    - D2 has more words - making it easier to get higher dot product and easier to get higher Euclidean distance.
    - D3 has fewer words - less likely to get higher dot product, more likely ot get a lower Euclidean distance. 

### Cosine

- Measures the angle between two vectors 
    - Normalized dot product



### Advantages of Cosine

- Symmetric 
- Bounded in [-1, 1]
    - cos(X, X) = 1
    - cos(X, -X) = -1
- Non-negative vectors (e.g., document vectors): cos >= 0
- Orthogonal vectors (90* apart): cos = 0
    - usually means two documents are unrelated

# Singular Value Decomposition (SVD)

### Singular Value Decomposition (SVD)

$$ A = U\Sigma V^T $$

- $A$ is an $n \times p$ matrix, $ UU^T = I, VV^T - I$, and $\Sigma$ is an $n \times p$ diagonal matrix with non-zero singular values
- Generalization of eigendecomposition to an arbitrary matrix  

### Singular values and singular vectors



- $\sigma_i$ are (non-negative) singular values of $A$, placed on the diagonal of $\Sigma$
- Every column of $U$ is a left singular vector
- Every column of $V$ is a right singular vector 

### SVD and eigendecomposition

- $ A = U\Sigma V^T$
- $U$ is the matrix of eigenvectors of $AA^T$
- $V$ is the matrix of eigenvectors of $A^T A$
- The eigenvectors are *orthogonal*. Why?
- Components of $\Sigma$ are the square roots off non-zero eigenvalues of $A^T A$ (and of $AA^T$)

### SVD for matrix tranformation 

- Unlike eigendecomposition, SVD exists for *all* real matrices, dense or sparse, square or not, normalized or not
- This is desirable for real world data!

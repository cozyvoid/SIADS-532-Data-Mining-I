# From Itemsets to Vectors

### Limitation of Itemsets

- Itemset representation ignores order and quanity
- It does not handle numerical data

### Example: shopping baskets

- {:beer:, :baby_bottle:, :watermelon:, :beer:}, and {:baby_bottle:, :beer:, :watermelon:} has the same itemset representation

- {:beer: 2, :baby_bottle: 1, :watermelon: 1} and {:beer: 1, :baby_bottle: 1, :watermelon: 1} would have been a better representation 

- What about {:beer:, :baby_bottle:, :lemon:}?

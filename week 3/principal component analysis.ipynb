# Principal Components Analysis

### Principal Components

- **Principal Components**: *orthogonal directions* that capture most of the variance in the data
    - The first principal component explains the most variance of the data
    - The next principal component explains most of the rest of the variance 

### Principal Component Analysis (PCA)

- Find the eigenvalues and eigenvectors of the *covariance matrix* 
    - Eigenvectors with the largest eigenvalues correspond to the dimensions that have the strongest correlation in the dataset 

### PCA

### Summary of PCA

- Principal components are special patterns of matrix data (orthogonal eigenvectors of the covariance matrix)
- PCA transforms original vectors into low dimensinal, orthogonal vector space
- Very usefful step of feature engineering in machine learning 

### Beyond PCA

- Eigenvalues and eigenvectors only apply to square matrices
- Stable eigendecomposition requires symmetric, positive semi-definite matrix 
- That's why we apply it to $X^T X$
- In reality, most data matrices are not square ($n! = p$)

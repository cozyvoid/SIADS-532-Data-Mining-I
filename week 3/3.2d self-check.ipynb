### 1. How does the Pearson Coefficient relate to cosine? (review)

The **Pearson correlation coefficient** is closely related to **cosine similarity**. Both measure how similarly two vectors vary across dimensions. The key difference is that Pearson first centers the data by subtracting each vector’s mean before comparing them.

So:

- **Cosine similarity** compares the angle between the original vectors
- **Pearson correlation** compares the angle between the mean-centered vectors

Because of this, Pearson measures whether two vectors move together relative to their own averages, while cosine measures similarity in overall direction without adjusting for each vector’s mean.

### 2. In what cases or types of data is using the Pearson Coefficient more meaningful? (review)

Pearson correlation is more meaningful when the important question is whether two variables have a **linear relationship**, rather than whether they simply have similar size or direction.

It is especially useful for:

- **ratings data**, where people may use different baseline scales
- **user preference data**, such as recommendation systems
- **relationships between variables**, such as income and spending, hours studied and exam score, or temperature and energy use
- cases where you care about whether two patterns rise and fall together, even if one is generally higher or lower than the other

For example, if two users rate movies similarly but one gives mostly high ratings and the other gives mostly low ratings, Pearson can still show that their preferences are aligned because it adjusts for their different rating averages.

### 3. Think about the different types of data that you could collect in the real world in vectors or matrices. For what kinds of data would using dot product, Manhattan distance, Euclidean distance, cosine, or Pearson's Correlation make more sense? It depends on what aspect of the data you want to capture:

- **Dot product** makes more sense when both **shared features and magnitude** matter.
    * Example: comparing weighted feature vectors where larger shared values should increase similarity.
- **Manhattan distance** makes more sense when you want to measure **total difference across dimensions** in a straightforward way.
    * Example: comparing two people’s spending across categories, where each category difference adds to the total.
- **Euclidean distance** makes more sense when **straight-line closeness** is meaningful.
    * Example: clustering points in physical space or comparing measurements where geometric distance is natural.
- **Cosine similarity** makes more sense when you care about **pattern or orientation**, not absolute size.
    * Example: document similarity using word-frequency vectors, where longer documents should not automatically seem more similar.
- **Pearson correlation** makes more sense when you care about whether two vectors have a similar **trend relative to their own averages**.
    * Example: comparing user ratings, stock movement patterns, or other data where baseline level differs across people or variables.

Overall:

- use **dot product** for shared magnitude
- use **Manhattan/Euclidean** for distance or dissimilarity
- use **cosine** for directional similarity
- use **Pearson** for linear relationship after adjusting for baseline differences

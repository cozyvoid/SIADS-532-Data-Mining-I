# Pearson Correlation Coefficient

### Pearson Coefficient

- Linear correlation between two variables
- Deducting mean from vector, then cosine



### Pearson vs Cosine

- Pearson coefficient is essentially cosine after deducting the mean from each vector
- "Mean" of vector indicates the trend
- Useful if dimensions are homogeneous (e.g., ratings on movies only)
- not meaningful if dimensions are hetergeneous (e.g., weight and height)

### Example: similarity / distance calculation 

- Three documents:
    - D1 contains "data mining"
    - D2 contains "data mining data mining"
    - D3 contains "data science"

- Dimensions: {data, mining, science}

    - $D1 = \langle 1, 1, 0 \rangle$
    - $D2 = \langle 2, 2, 0 \rangle$
    - $D3 = \langle 1, 0, 1 \rangle$

- **Dot Product:** 
    - $ sim(D1, D2) = 1 * 2 + 1 * 2 + 0 * 0 = 4 $
    - $ sim(D1, D3) = 1 * 1 + 1 * 0 + 0 * 1 = 1 $

- **Manhattan Distance:**
    - $ d(D1, D2) = |1-2| + |1-2| + |0-0| = 2 $
    - $ d(D1, D3) = |1-1| + |1-0| + |0-1| = 2 $

- **Euclidean Distance:** 
    - $ d(D1, D2) = \sqrt {(1-2)^2 + (1-2)^2 + (0-0)^2} = 1.4 $
    - $ d(D1, D3) = \sqrt {(1-1)^2 + (1-0)^2 + (0-1)^2} = 1.4 $ 

- **Cosine:**
$$\cos(D1, D2) = \frac{1 * 2 + 1 * 2 + 0 * 0}{\sqrt{1^2 + 1^2 + 0^2} \cdot \sqrt{2^2 + 2^2 + 0^2}} = \frac{4}{\sqrt{2} \cdot \sqrt{8}} = 1$$
$$\cos(D1, D3) = \frac{1 * 1 + 1 * 0 + 0 * 1}{\sqrt{1^2 + 1^2 + 0^2} \cdot \sqrt{1^2 + 0^2 + 1^2}} = \frac{1}{\sqrt{2} \cdot \sqrt{2}} = 0.5$$

# Eigendecomposition

### how are these patterns useful?

- Eigenvectors are the basis of many matrix decomposition operations
- Matrix decomposition (matrix factorization): factorize a matrix into a product of matrices

### Eigendecomposition



- $A$ is a square matrix with distinct eigenvalues, $\Lambda$ is a diagonal matrix, and $UU^-1 = 1$



- $A$ is a square matrix
- $\Lambda$ is a diagonal matrix with $n$ eigenvalues
- $U$ is a matrix off eigenvectors (each column of $U$ is the corresponding eigenvector)

### Other facts about eigendecomposition 

- Eigendecomposition doesn't exist for all square matrices
- It applies to a "diagonalizable" matrix, which much have $n$ distinct eigenvalues
- Always applies to a *symmetric* matrix
- Eigenvalues are non-negative for a "positive semidefinite" matrix

### This is good

- But it still doesn't tell uis why it is useful in reality...

### Issues of vector representation

- In reality, there are too many possible dimensions
- Dimensions are highly correlated
    - Some are even redundant
    - Seriously affects vector similarity/distance!

### Example: temperatures of 2 cities

- Day avg. and Night avg. (in fahrenheit):
    - City A: $ \langle 77, 59 \rangle $
    - City B: $ \langle 68, 50 \rangle $
    - d(A, B) = 12.7

Intuitively, difference between City A and City B is 9*F
**Issue**: day and night temps are highly correlated

- Add two dimensions (in celsius):
    - City A: $ \langle 77, 25, 59, 15 \rangle $
    - City B: $ \langle 68, 20, 50, 10 \rangle $
    - d(A, B) = 14.6

**Issue**: highly correlated and redundant dimensions

### Intuitive solution

- How can we construct dimensions that are compact and orthogonal to each other?
- Matrix decompostion helps!

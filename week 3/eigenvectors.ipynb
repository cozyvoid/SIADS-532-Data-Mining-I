### Eigenvectors and eigenvalues

An eigenvector is an implicit "direction" for a square matrix.



### Computing eigenvalues

Computing eigenvalues of a square (n x n) matrix (det = determinant):

$$ det(A - \lambda I) = 0 $$

- $r$ distinct solutions, where $ l <= r <= N$
- For each eigenvalue ($\lambda$) found, you can find an eigenvector ($v$)

### Properties of eigenvectors 

- Eigenvectors are invariant directions of a linear transformation (defined by matrix $A$)
- Eigenvectors of distinct eigenvectors are linearly independent to each other
- Eigenvectors of a *symmetric* matrix are orthogonal to each other!

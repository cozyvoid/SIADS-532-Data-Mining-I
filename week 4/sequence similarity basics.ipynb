# Sequence Similarity Basics

---

### Sequence Representation of Data

- Sequence = items + order
    - Order matters
    - Repeating items matter
    - Absolute positions of items don't matter

### Challenges of Computing Sequences Similarities

- First try: Flattening



- Flatten into itemset: $ {x_1, x_2,..., x_k} $
- Or flatten into a vector of items
- Then similarity can be calculated using itemset/vector similarity

### Problem with flattening

- Order is a big factor in sequence similarity
    - e.g., "live" vs. "evil"
- Both itemsets and item vectors fail to handle order

#### Second Try: Hamming Distance

- Equal-length sequences: number of fpositions at which the corresponding items are different
    - e.g., distance between "Carolyn" and "Karolin" is 2 
        (differences in teh same positions: 1 and 6)

**What about "awake" and "waked"?**

### Sequence similarity: intuition

- The more items in common, the more similar
- The more aligned the order, the more similar

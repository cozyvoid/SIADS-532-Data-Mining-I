### 1. What type of information is lost when you flatten a sequence? Why is this a problem? (review)

When you **flatten a sequence**, you usually lose the **order of events**.

For example, this sequence:

`<A, B, C>`

might be flattened into the itemset:

`{A, B, C}`

The flattened version tells us that A, B, and C all occurred, but it does **not** tell us that A came before B, or that B came before C.

This is a problem because many real-world patterns depend on order. For example:

`<View Product, Add to Cart, Purchase>`

means something very different from:

`<Purchase, View Product, Add to Cart>`

If we flatten both into:

{View Product, Add to Cart, Purchase}

then they look identical, even though the behavior is very different.

Flattening can make sequence data easier to analyze with itemset methods, but it removes the temporal or ordered structure that often contains the most important meaning.

### 2. What aspect of data does Hamming Distance measure that sequences don't measure (review)? Why is this a problem? (review)

**Hamming Distance** measures differences at **specific positions** between two sequences of the same length.

For example:

`A B C D`
`A B E D`

Only one position is different:

- Position 1: A = A
- Position 2: B = B
- Position 3: C ≠ E
- Position 4: D = D

So the Hamming Distance is 1.

This means Hamming Distance cares about **exact positional alignment**. A value in position 3 is compared only to the value in position 3 of the other sequence.

This is different from many sequence mining tasks, where the main concern is usually **relative order**, not exact position.

For example:

`<A, B, C, D>`
`<A, X, B, C, D>`

These sequences are very similar in order, because both contain A, then B, then C, then D. But Hamming Distance does not handle this well because the second sequence has an extra item, shifting the positions. It either cannot be applied directly because the lengths differ, or it would compare misaligned positions.

This is a problem because real-world sequences often have insertions, deletions, skipped actions, or extra noise. Hamming Distance is too rigid when sequences are not perfectly aligned.

### 3. What aspects of data does Levenshtein Edit Distance measure that isn't captured in the previous methods? (review)

**Levenshtein Edit Distance** measures how many small edits are needed to transform one sequence into another.

It captures differences caused by:

- **Substitution**: one item is replaced by another.
- **Insertion**: an extra item appears.
- **Deletion**: an item is missing.

This makes it more flexible than Hamming Distance because it can compare sequences of different lengths.

For example:

`<A, B, C, D>`
`<A, X, B, C, D>`

The second sequence has one extra item, `X`.

Using Levenshtein Edit Distance, we can say the distance is `1` because we only need one insertion:

`<A, B, C, D> → <A, X, B, C, D>`

This captures the idea that the sequences are still very similar even though one has an extra event.

Levenshtein Edit Distance captures **sequence similarity with alignment flexibility**. It recognizes that two sequences may be similar even if items are shifted because of missing or extra elements.

### 4. What unit operations does Levenshtein Edit Distance use? What other unit operations do other edit distance methods use, and how does that change what they measure? (review)

Levenshtein Edit Distance uses three unit operations:

1. **Insertion**: add an item.
2. **Deletion**: remove an item.
3. **Substitution**: replace one item with another.

Each operation usually has a cost of 1.

For example, to transform:

`cat → cut`

We substitute `a` with `u`.

So the Levenshtein distance is `1`.

To transform:

`cat → cart`

We insert `r`.

So the distance is also `1`.

Other edit distance methods may allow different operations. For example, **Damerau-Levenshtein Distance** also allows:

- **Transposition**: swapping two adjacent items.

For example:

`form → from`

The `o` and `r` are swapped.

With regular Levenshtein Distance, this may require two substitutions:

`form → from`

But with Damerau-Levenshtein Distance, it can be treated as one transposition.

This changes what the distance measures. If transpositions are allowed, the method is better for data where swaps are common, such as typing errors.

Other methods may use:

- **Different operation costs**, where some edits are considered more serious than others.
- **Gap penalties**, often used in biological sequence alignment, where long gaps may be treated differently from many separate insertions or deletions.
- **Match/mismatch scores**, where similarity is scored rather than only counted as edit operations.

The choice of unit operations affects the meaning of similarity. For example:

- Hamming Distance emphasizes exact position-by-position differences.
- Levenshtein Distance emphasizes insertions, deletions, and substitutions.
- Damerau-Levenshtein Distance also recognizes swaps.
- Gap-based methods emphasize aligned regions and interruptions.

So edit distance methods differ based on what kinds of sequence changes they consider meaningful.

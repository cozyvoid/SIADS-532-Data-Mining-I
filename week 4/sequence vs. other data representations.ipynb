# Sequence vs. Other Data Representations

---

### Sequence vs. itemset / vector

- itemset: {:beer:, :baby_bottle:, :watermelon:, :beer:} = {:beer:, :baby_bottle:, :watermelon:}, = {:baby_bottle:, :beer:, :watermelon:}
- vector: {:beer:, :baby_bottle:, :watermelon:, :beer:} $\neq$ {:beer:, :baby_bottle:, :watermelon:} = {:baby_bottle:, :beer:, :watermelon:}
- sequence: {:beer:, :baby_bottle:, :watermelon:, :beer:} $\neq$ {:beer:, :baby_bottle:, :watermelon:} $\neq$ {:baby_bottle:, :beer:, :watermelon:}

**Sequence:**
- Order does matter
- Repeated occurrence of item does matter 

### Sequence of vectors

A generalization from sequence of itemset



- $x_k$ is a vector of $p$ dimensions
- $p$ numerical attributes to describe an item
- $p$ dimensions are aligned in $x_i, x_2, ..., x_k$

### Example: sequence of vectors

- Trajectory -> sequence of locations;
- Every location as a 2D vector 
    - longtitude
    - latitude 


[2002 Winter Olympics Torch Relay Route]

### Sequence of vectors vs. matrix 

- Seqeuence of vectors is closely related to a $k \times p$ matrix (or $p \times k$ after transpose)



### Sequence of vectors vs. matrix 

- Sequence of vectors is closely related to a $k \times p$ matrix (or $p \times k$ after transpose)
    - But the rows ($k$ objects) now have an order and cannot be rearranged
    - Must be careful when applying PCA style dimension reduction

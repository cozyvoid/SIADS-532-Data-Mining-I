# Sequence Similiarity through Shingling

---

### Application: near-duplicates



- 30% of web pages are near-duplicates (Broder et al. 1997)
    - mirrors, local copies, page updates
    - spam, spider-traps
    - crawling errors
- Detect near-duplicates and remove them from index (with limited time & space cost)

### Challenge 

- Strict duplication is easy to handle (use hashing), but what about near-duplicaton?
- Similar problem: **plagarism detection**
- Need a fast measurement of sequence similarity (for long sequences)

### Shinglong (Broder at al., 1997)

- Represent a sequence as a set off overlapping ngrams
    - "to be or not to be" -> 
        - {"to be or", 
            - "be or not",
                - "or not to",
                    - "not to be"}

- Then sequence similarity is Jaccard similarity of two sets

$$
J (A, B) = \frac {|A \cap B|}{|A \cup B|} 
$$

### Example

- X = "to be or not to be"
- Y = "not be or not to be"

    - X': {"to be or", "be or not", "or not to", "not to be"}
    - Y': {"not be or", "be or not", "or not to", "not to be"}

- $sim(X,Y) = J(X', Y') = 3/5 $

### Example

- X = "to be or not to be"
- Y = "not be or not to be"
- Z = "be or not to not be"

    - X': {"to be or", "be or not", "or not to", "not to be"}
    - Y': {"not be or", "be or not", "or not to", "not to be"}
    - Z': {"be or not", "or not to", "not to not", "to not be"}

- $sim(X, Y) = J(X', Y') = 3/5 $
- $sim(X, Z) = J(X', Z') = 2/6 $

### Advantages of Shingling

- Fast: O(m + n) to compare two sequences
- Bounded between [0,1]
- Can be further speeded up using random projection algorithms
    - Minhash
    - Locality sensitive hashing

### Limitations of Shingling

- Set approximation may not be accurate
- Size of shingles matter
- Variations of algorithms try to estimate other similarity functions (e.g., Weighted Jaccard, Euclidean distance)

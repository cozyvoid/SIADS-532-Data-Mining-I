# Subsequences

---

- $X'$ is a subsequence of $X$ after dropping items in $X$ and keeping the order



- All itemms in $X'$ are part of the original sequence $(X)$:
    - $n_j$ is a number among $1,2, ..., k$

### Example of subsequences

## "to be or not to be"

- e.g., subsequences length 3:
    - "to be or", "be or not", "or not to", "not to be"     (consecutive sequence)
    
    - "to be not", "be not be", "to to be", "to or not", ...    (non-consecutive subsequence)

    - "not be to", "or be be"

### How to solve the sub-problems?

- How do you find:
    - $d("counter", "compute")$
    - $d("counte", "computer")$
    - $d("counte", "compute")$
- Recursively back off to shorter subsequences, until we know the answer.

### What do we know for sure?

- $d(X, "") = length(X)$
- $d("", Y) = length(Y)$

- Example: 
    - $d("counter" -> "") = 7$
    - $d("" -> "computer") = 8$
    - $d("c" -> "") = 1$

- Start from there and grow into full sequences.

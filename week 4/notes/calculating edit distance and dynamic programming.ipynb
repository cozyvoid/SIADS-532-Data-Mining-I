# Calculating Edit Distance and Dynamic Programming

---

### How to calculate edit distance

- Easy to find transformations, but difficult to find one with the "smallest" number of edits



- How can we calculate which transformation is the smallest?

### Solution: Dynamic Programming

- Dynamic programming breaks down a complex problem into simpler sub-problems in a recursive manner
    - Store solutions for sub-problems
    - Use these solutions to find the solution of a complex problem

### Edit distance as dynamic programming

- Complex problem: Find the edit distance of two sequences.
- Sub-problems: Find the edit distance of two (shorter) sub-sequences.

### Example

- Let $d(X, Y)$ be the edit distance between sequences $X$ and $Y$.
    - We can find $d("counter", "computer") if we know all of the following:
        - $d("counter", "compute")$
        - $d("counte", "computer")$
        - $d("counte", "compute")$

# Frequent Sequential Patterns

---

### Frequent sequential patterns

- sequence = catergorical items + order
- Frequent sequential patterns = frequent itemsets + order 
    - Repeating items matter

### Example: shopping baskets



- {:beer:}: support = 80% 
- {:beer:, :baby_bottle:}: support = 60%
- {:beer:, :baby_bottle:, :lemon:}: support = 40%

> If $min_sup = 50%, then {:beer:} and {:beer:, :baby_bottle:} are frequent.

### Example: shopping sequences



- {:beer:}: support = 80%  :white_check_mark:
- {:beer:, :baby_bottle:}: support = 60% :x:, 40%
- {:beer:, :baby_bottle:, :lemon:}: support = 40% :x:, 20%

> If $min_sup = 50% $, then only [:beer:] is frequent.

### Apriori criterion still holds

- If a sequence is frequent, all its subsequences have to be frequent.
- {:beer:, :baby_bottle:} is not frequent -> no need to check {:beer:, :baby_bottle:, :lemon:}

### From itemsets to sequences

- Generalized Apriori algorithm works on sequences (Agrawal and Strikant, EDBT'96)
    - Association rules still applies, but order sensitive
    - Pattern  evaluation metrics still apply, but order sensitive

### Association rule without order



- {:beer:, :baby_bottle:}: support = 60%
- {:beer:} -> {:baby_support:}: support 60%, confidence = 75%

### Association rule with order



- {:beer:, :baby_bottle:}: support = 40%
- {:beer:} -> {:baby_bottle:}:
    - support = 40%
    - confidence = 50%
- [40%, 50%]

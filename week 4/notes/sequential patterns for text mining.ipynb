# Sequential Patterns for Text Mining

---

### Sequential patterns and text mining

- Text data are naturally represented as sequences
    - Item = word or character
    - Sequential pattern = phrases, collocations, or expressions

### Extracting N-grams and Skip-grams in text

- **N-gram examples**
    - "Ann Arbor"
    - "hot dog"
    - "in front of"
    - "Txt U L8R"

- **Skip-gram examples**
    - "not only... but also"
    - "deep * networks"

### Example of sequential patterns for text mining

- **"to be or not to be"**
    - Represented as set of trigrams:
        - {"to be or", "be or not", "or not to", "not to be"}
    - Represented as a vector of biograms:
        - Dimensions = {"to be", "be or", "or not", "not to"}
        - Vector = $ \langle 2, 1, 1, 1 \rangle $
    - Then all functionalities for itemset/vectors can be applied

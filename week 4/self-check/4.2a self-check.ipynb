### 1. 1. How does a sequence of itemsets differ from using itemsets or sequences on their own?

A sequence of itemsets combines two ideas:

- **Itemsets** capture items that occur together.
- **Sequences** capture the order in which events occur.

A regular itemset ignores order. For example:

`{milk, bread, eggs}`

This tells us the items occurred together, but not when each item appeared.

A regular sequence captures order, but usually treats each position as a single item. For example:

`<milk, bread, eggs>`

This tells us milk came before bread, and bread came before eggs.

A **sequence of itemsets** lets each step contain multiple items:

`<{milk, bread}, {eggs}, {coffee, sugar}>`

This means:

- milk and bread occurred together first,
- then eggs occurred later,
- then coffee and sugar occurred together later.

This is useful when data has both **co-occurrence** and **order**. For example, in shopping data, a customer may buy several items during one visit, then return later and buy another group of items. In web data, a user may click several related links during one session, then move to another group of actions later.

So a sequence of itemsets is more expressive because it captures **what happened together** and **what happened next**.

### 2. Why does the Apriori Algorithm still work on finding sequences of itemsets? (review)

Apriori still works because sequence patterns also follow the **Apriori principle**, also called the **downward-closure property**.

The Apriori principle says:

> If a pattern is frequent, then all of its smaller subpatterns must also be frequent.

For itemsets, this means if `{milk, bread, eggs}` is frequent, then `{milk, bread}, {milk, eggs}`, and `{bread, eggs}` must also be frequent.

The same idea applies to sequences of itemsets.

For example, suppose this sequence pattern is frequent:

`<{milk, bread}, {eggs}>`

Then its smaller subsequences must also be frequent, such as:

`<{milk}, {eggs}>`

`<{bread}, {eggs}>`

`<{milk, bread}>`

`<{eggs}>`

Because every sequence that contains the larger pattern must also contain its smaller pieces.

This lets Apriori prune the search space. If a smaller sequence is not frequent, then any larger sequence containing it cannot be frequent either.

For example, if:

`<{milk}, {eggs}>`

is not frequent, then we do not need to test larger patterns such as:

`<{milk, bread}, {eggs}>`

or

`<{milk}, {eggs, coffee}>`

because they cannot be frequent.

So Apriori still works because sequence patterns preserve the same logic: **larger frequent patterns require smaller frequent patterns**.

### 3. Make sure you understand the difference between n-grams and skip-grams. Can you come up with common examples used in natural language besides the examples in the lectures? (review)

An **n-gram** is a sequence of `n` consecutive words, characters, or tokens.

For example, in the sentence:

`I love data mining`

The word bigrams are:

`I love`
`love data`
`data mining`

The word trigrams are:

`I love data`
`love data mining`

N-grams are useful because they capture local word order and common phrases.

A **skip-gram** allows words to be skipped between the selected tokens.

Using the same sentence:

`I love data mining`

Some skip-grams could be:

`I data`
`I mining`
`love mining`

These are not consecutive words, but they still preserve order.

The key difference is:

- **N-grams**: tokens must be next to each other.
- **Skip-grams**: tokens can have gaps between them.

Common natural language examples:

In the phrase:

`New York City subway`

Some n-grams are:

- `New York`
- `York City`
- `City subway`
- `New York City`

These capture fixed phrases and named entities.

Some skip-grams are:

- `New City`
- `New subway`
- `York subway`

These may help detect related words even when other words appear between them.

Another example:

`The movie was not very good`

N-grams:

- `not very`
- `very good`
- `not very good`

Skip-grams:

- `not good`
- `movie good`
- `was good`

The skip-gram `not good` is especially useful because it captures the negative sentiment relationship even though the word `very` appears between them.

Another example:

`The patient was diagnosed with high blood pressure`

N-grams:

- `high blood`
- `blood pressure`
- `high blood pressure`

Skip-grams:

- `diagnosed high`
- `diagnosed pressure`
- `patient pressure`

In text mining, n-grams are good for capturing exact phrases, while skip-grams are useful for finding looser relationships where important words appear near each other but not necessarily side by side.

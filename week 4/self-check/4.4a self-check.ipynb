### 1. Shingling detects which particular aspect of real world datasets? (review)

**Shingling detects local overlap in ordered data.**

A **shingle** is a small consecutive chunk of a sequence, such as a group of words, characters, or events. For example, if we use word shingles of size 3 on this sentence:

`data mining finds useful patterns`

The 3-word shingles are:

- `data mining finds`
- `mining finds useful`
- `finds useful patterns`

Shingling is useful because it captures **near-duplicate or highly similar content** without requiring the entire sequence to match exactly.

In real-world datasets, shingling is often used to detect things like:

- Similar documents
- Plagiarized text
- Repeated web pages
- Near-duplicate product descriptions
- Similar DNA or protein sequence fragments
- Similar clickstream or behavior patterns

The key idea is that two objects are similar if they share many of the same shingles.

For example:

`The customer bought milk and bread yesterday`

and

`The customer bought milk and bread today`

share many overlapping word shingles, so shingling would detect that the two sentences are highly similar even though they are not exactly identical.

So shingling detects **similarity based on shared local sequence patterns**.

### 2. Why is the run time of shingling faster than Levenshtein Edit Distance? [This isn't directly described in the lecture. Hint: Focus on how the dynamic programming aspect of calculating edit distance compares to using Jaccard Similarity] (review)

Shingling is usually faster because it turns sequence comparison into a **set similarity problem**, often using **Jaccard Similarity**, instead of solving a full alignment problem.

Levenshtein Edit Distance compares two sequences by asking:

> What is the minimum number of insertions, deletions, and substitutions needed to transform one sequence into the other?

To answer that, it typically uses **dynamic programming**. If one sequence has length `m` and the other has length `n`, the algorithm fills an `m × n` table.

So the runtime is usually:

`O(mn)`

That means comparing long sequences can become expensive because every position in one sequence is compared against many positions in the other sequence.

Shingling works differently. It breaks each sequence into chunks, then compares the sets of chunks.

For example:

Sequence 1:

`A B C D E`

3-shingles:

`ABC, BCD, CDE`

Sequence 2:

`A B C X E`

3-shingles:

`ABC, BCX, CXE`

Then we compute overlap:

- Intersection: `{ABC}`
- Union: `{ABC, BCD, CDE, BCX, CXE}`

Jaccard Similarity:

`1 / 5`

This avoids building a full dynamic programming table. Instead, the work is mostly:

1. Generate shingles.
2. Store them in sets.
3. Count intersection and union.

With hash sets, this is usually much closer to linear in the number of shingles.

So the basic runtime is closer to:

`O(m + n)`

for one pairwise comparison, assuming efficient hashing.

The tradeoff is that shingling is faster but less exact. Levenshtein Edit Distance gives a precise edit-based alignment distance, while shingling gives an approximate similarity based on shared local chunks.

In short:

- **Levenshtein Edit Distance** is slower because it performs dynamic programming over pairwise positions.
- **Shingling** is faster because it converts sequences into sets and compares overlap using Jaccard Similarity.
- **Levenshtein** measures edit operations.
- **Shingling** measures shared local patterns.

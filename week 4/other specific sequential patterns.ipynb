# Other Specific Sequential Patterns

--- 

## Special cases of sequential patterns

- N-grams
    - Subsequences are consecutive (relative to original sequences)

- **"to be or not to be"**
    - 1-grams (unigrams): "to", "be", "or", "not"
    - 2-grams (bigrams): "to be", "be or", "or not", "not to"
    - 3-grams (trigrams): "to be or", "be or not", "or not be", "not ot be"
    - 4-grams: "to be or not", "be or not to", "or not to be"

### Skip-grams

- Subsequences have gaps (relative to original sequences)

- **"to be or not to be"** (use astericks to indicate gaps)
    - Skip-bigrams: "to * be", "to * or", "to * not", "to * to"
    - Skip-trigrams: "to be * not", "to * or to", "to * to be", ...

### N-grams

- Subsequences are consecutive
- n = natural number
    - n = 1: 1 item
    - n = 2: 2 consecutive items
    - n = 3: 3 consecutive items
    - ...

### Skip-grams

- Subsequences have gaps
- Number of skipped items in gap is arbitrary
- Types:    
    - Skip-bigram = 2 item subsequence with gaps
    - Skip-trigram = 3 item subsequence with gaps 

### K-Skip-grams

- Subsequences with gaps skipping at most K items in the original sequence

- **"to be or not to be"**

- 2-skip-trigrams:
    - "to * or not", "to * not * be"    :x:
    - "to * to be", "to be * be"        :x:

### Prefix and Suffix

- **Prefix**: subsequence that contains the first items of a sequence
    - "sub" in "subsequence"
- **Suffix**: subsequence that contains the last items of a sequence
    - "less" in "priceless"

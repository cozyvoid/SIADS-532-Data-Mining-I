### 1. What major aspect of data do sequences measure that both itemsets and vectors do not? (introduction to sequences, or an example of how they differ)

Sequences measure **order**.

An **itemset** tells us what items appear together, but not the order in which they appear. A vector tells us values across fixed dimensions, but it also usually does not represent a before/after relationship unless time or order is explicitly built into the features.

For example:

- Itemset: `{bread, milk, eggs}`
    - This only says the items occurred together. It does not say whether someone bought bread before milk.
- Vector: `[5, 2, 9]`
    - This gives values in fixed positions, but the positions represent features, not necessarily events over time.
- Sequence: `<bread, milk, eggs>`
    - This says bread occurred first, then milk, then eggs.

So the main thing sequences capture is **the ordering of events, actions, words, purchases, clicks, locations, or states**. This is important when the pattern depends on what happens before or after something else.

### 2. Does the exact position of a data point in a sequence matter, or is it just the order that matters? (review and a walked-through example)

In most sequence mining tasks, the **relative order matters more than the exact position**.

For example, consider this sequence:

`<A, B, C, D, E>`

The pattern `<B, D>` appears in the sequence because **B comes before D**. It does not matter that B is in position 2 and D is in position 4 unless the problem specifically cares about exact timing or exact index.

Now compare:

`<A, C, B, D, E>`

The pattern `<B, D>` still appears because B still comes before D.

But in this sequence:

`<A, D, C, B, E>`

The pattern `<B, D>` does not appear because D comes before B. The order is reversed.

So generally:

- Exact position: usually not the main focus.
- Relative order: very important.
- Exact position only matters if the task defines it as important, such as “the event must happen exactly third” or “within two time steps.”

### 3. Does the exact position of a data point in a subsequence matter? (review, can continue to 1:40 for an example of how gaps are handled in subsequences)

Usually, no. In a subsequence, the **relative order matters**, but the exact positions do not have to match.

A subsequence can allow **gaps** between elements.

For example, consider the full sequence:

`<A, B, C, D, E>`

The subsequence `<B, D>` is present because B appears before D, even though C appears between them.

Here, the gap is:

`B, [C], D`

The exact positions are:

- B is at position 2
- D is at position 4

But the subsequence `<B, D>` does not require B and D to be next to each other. It only requires that B occurs before D.

However, `<D, B>` is not a subsequence of `<A, B, C, D, E>` because D does not occur before B.

So for subsequences:

- The exact position usually does not matter.
- The order must be preserved.
- Gaps are allowed unless the problem specifically requires consecutive items.

### 4. How is using a sequence of vectors different than a matrix? Is there a way to represent a sequence of vectors as a matrix? (review)

A **sequence of vectors** means each step in the sequence is represented by multiple features.

For example, suppose we track a person’s exercise over several days:

- `Day 1: [distance, heart rate, calories] = [3.0, 140, 300]`
- `Day 2: [distance, heart rate, calories] = [4.5, 150, 420]`
- `Day 3: [distance, heart rate, calories] = [2.0, 130, 220]`

This can be written as a sequence of vectors:

`<[3.0, 140, 300], [4.5, 150, 420], [2.0, 130, 220]>`

This can also be represented as a matrix:

| Time step	| Distance	| Heart rate |	Calories |
|---|---|---|---|
| 1	|3.0 |	140 |	300 |
| 2	| 4.5	| 150	| 420 |
| 3	| 2.0	| 130	| 220 |

So yes, a sequence of vectors can be stored as a matrix, where:

- Rows represent time steps or sequence positions.
- Columns represent features.

The difference is mainly in how we interpret it. A regular matrix might just be a table of values, while a sequence of vectors treats the **row order as meaningful**. The first row happens before the second, and the second happens before the third.

So the matrix representation is possible, but we must remember that the rows are ordered events, not independent records.

### 5. How does the use of gaps affect the use of sequences with regards to finding patterns? (review)

Gaps make sequence mining more flexible because patterns do not have to appear as consecutive items.

For example, suppose we are looking for the pattern:

`<A, C>`

Now consider these sequences:

`<A, C, D>`
`<A, B, C, D>`
`<A, B, D, C>`

The pattern `<A, C>` appears in all three because A occurs before C. The extra items between them are allowed as gaps.

This is useful because real-world behavior often contains noise or extra actions. For example, in web browsing, a user might:

`<View Product, Read Reviews, Compare Prices, Add to Cart>`

The important pattern might be:

`<View Product, Add to Cart>`

Even though other actions happened in between, the overall behavioral pattern is still meaningful.

However, allowing gaps can also make pattern mining harder because many more possible subsequences may be found. This can increase the number of patterns and may include weak or unimportant patterns.

So gaps have two main effects:

- They help find broader, more flexible patterns.
- They can also create more candidate patterns, making it harder to identify which patterns are truly meaningful.

In short, gaps allow sequence mining to capture realistic patterns, but they also require careful filtering so the results are not too noisy.

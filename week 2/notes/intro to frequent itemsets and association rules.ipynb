# Introduction to Frequent Itemsets and Association Rules

### What is a "pattern"?

- A structure of attributes that represents the intrinsic and important properties of data objects

### A "pattern" in itemset data

- A **frequent subset** of itemset
- An **association rule**
- A **correlation** of two itemset


### Subset and Superset

Two itemsets $X_1$ and $X_2$:

If every item in $X_1$ is also in $X_2$, we say:

- $X_1$ is a **subset** of $X_2$ ($X_1$ appears in $X_2$)
- $X_2$ is a **superset** of $X_1$ (X_2$ contains $X_1$)

{:watermelon:, :rabbit:} is a subset of {:rabbit:, :ram:, :peach:, :watermelon:, :orange:}

### Frequent itemset

A k-itemset: $X = {x_1, x_2, ..., x_k}$

Support: frequency of $X$ in a database:

- **Absolute support**: number of transactions that contain $X$
- **Relative support**: fraction of transaction that contain $X$

An itemset $X$ is frequent if its support $sup(X)$ is no less than a threshold $ min_sup $

### Example: shopping baskets

| TID | Items Bought |
|-----|-----|
| 1 | :beer:, :baby_bottle:, :watermelon: |
| 2 | :beer:, :lollipop:, :baby_bottle:, :lemon: |
| 3 | :baby_bottle:, :beer:, :lemon: |
| 4 | :baby_bottle:, :lollipop: |
| 5 | :baby_bottle:, :lollipop: |
| 6 | :wine_glass:, :cookie:, :beer:, :bread:, :lemon: |

* {:beer:}: support = 80%
* {:beer:, :baby_bottle:}: support = 60%
* {:beer:, :baby_bottle:, :lemon:}: support = 40%

If min_sup = 50%, then {:beer:} and {:beer:, :baby_bottle:} are frequent.

### Association rules

$ X -> Y $

- Both $X$ and $Y$ are itemsets
- **Support**: probability that a transaction contains both $X$ and $Y$
- **Confidence**: conditional probability that a transaction that contains $X$ also contains $Y$

### Example

| TID | Items Bought |
|-----|-----|
| 1 | :beer:, :baby_bottle:, :watermelon: |
| 2 | :beer:, :lollipop:, :baby_bottle:, :lemon: |
| 3 | :baby_bottle:, :beer:, :lemon: |
| 4 | :baby_bottle:, :lollipop: |
| 5 | :baby_bottle:, :lollipop: |
| 6 | :wine_glass:, :cookie:, :beer:, :bread:, :lemon: |

* {:beer:, :baby_bottle:}: support = 60%
* {:beer:} -> {:baby_bottle:}:
    - support = 60%
    - confidence = 75%
    - [60%, 75%]
* {:beer:, :baby_bottle:} -> {:lemon:}:
    - [40%, 66.7%]

### 1. The definition of pattern is different depending on the type of dataset you are looking at. What are considered patterns in itemsets, and why? (review)

---

In itemset data, a **pattern** is usually a **set of items that appear together across many transactions or records**. These are often called **frequent itemsets**.

For example, in market basket data, if many customers buy **bread and milk** together, then 
`
{bread, milk}
` 
can be considered a pattern.

They are considered patterns because itemset data focuses on **co-occurrence**. The main question is not the order of items or their exact numeric values, but rather **which items tend to appear together**. A repeated combination suggests something meaningful in the data, such as shared behavior, preference, or correlation.

### 2. How do subsets and supersets help define patterns in itemset data? (review)

---

Subsets and supersets are important because itemset patterns are built from smaller and larger combinations of items.

A **subset** is a smaller part of an itemset.

Example: 
`
{bread, milk}
`
is a subset of 
`
{bread, milk, eggs}
`.

A **superset** is a larger itemset containing a smaller one.

Example: 
`
{bread, milk, eggs}
` 
is a superset of 
`
{bread, milk}
`.

These help define patterns because of the **downward closure property** in frequent itemset mining:

If an itemset is **frequent**, then **all of its subsets must also be frequent**.
If an itemset is **not frequent**, then **all of its supersets must also be not frequent**.

This matters because it helps algorithms prune the search space efficiently. If a small itemset is already too rare, there is no need to test bigger itemsets containing it.

### 3. What are absolute and relative support for frequent itemsets, and what is the threshold of support that defines a frequent itemset? (review)

---

**Support** tells us how often an itemset appears in the dataset.

**Absolute support** is the **raw count** of transactions containing the itemset.
Example: if 
`
{bread, milk} 
`
appears in 25 transactions, its absolute support is **25**.

**Relative support** is the **proportion** or percentage of transactions containing the itemset.
Example: if there are 100 total transactions and 
`
{bread, milk} 
`
appears in 25 of them, its relative support is **25/100 = 0.25**, or **25%**.

An itemset is considered **frequent** if its support is greater than or equal to a user-specified minimum support threshold, often called **min_sup**.

So the threshold is not universal. It depends on what the analyst sets. For example:

- If **min_sup** = **20 transactions**, then an itemset with absolute support 25 is frequent.
- If **min_sup** = **0.20**, then an itemset with relative support 0.25 is frequent.

### 4. How are association rules different than frequent itemsets? What is the particular feature by which association rules are measured that makes them different than frequent itemsets? (review)

---

A **frequent itemset** is simply a group of items that appears together often.

An **association rule** goes further by expressing a directional relationship:

$ X → Y $

This means: when `X` appears, `Y` tends to appear too.

Example:
`
{bread} → {milk}
` 
means customers who buy bread also tend to buy milk.

The key difference is that frequent itemsets are measured mainly by support, while association rules are especially measured by confidence.

- **Support** tells how often the whole rule occurs in the dataset.
- **Confidence** tells how often `Y` appears given that `X` appears.

So confidence is what makes association rules different, because it evaluates the strength of the implication rather than just the frequency of co-occurrence.

In simple terms:

- **Frequent itemset** = “These items often occur together.”
- **Association rule** = “If this item or set of items appears, that other item is likely to appear too.”

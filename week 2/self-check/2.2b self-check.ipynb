### 1. What is $χ2$ and how do you calculate it? How is it different from using Lift or Association Rules? (review)

---

$x^2$ (chi-square) is a statistical measure used to test whether two variables are **independent** or **associated**. In data mining, it is often used to check whether the occurrence of item 𝑋 is related to the occurrence of item `Y`, rather than happening together just by chance.

To calculate $x^2$, you compare the **observed counts** in the data to the **expected counts** you would get if the two items were independent.

The formula is:
$
x^2 = \sum \frac {(O - E)^2}{E}
$

where:
- O = observed count
- E = expected count under independence 

For two binary items, you usually organize the data into a 2 x 2 contingency table:
- `X` and `Y`
- `X` and not `Y`
- not `X` and `Y`
- not `X` and not `Y`

Then you compute the expected count for each cell as:
$
E = \frac {(row total)(column total)}{grand total}
$
and plug those values into the chi-square formula.

**How it differs from Lift and Association Rules**:
- Association rules describe directional relationships such as `X → Y`
- Lift measures how much more often `X` and `Y` occur together than expected if they were independent.
- $x^2$ tests whether the association is statistically meaningful by comparing observed and expected frequencies across all states

So:
- Association rules are about rule structure and prediction 
- Lift is about strength of co-occurence relative to chance
- $X^2$ is about statistical dependence


Another important difference is that lift is usually a **ratio**, while $x^2$ is based on **deviations between observed and expected counts**.

### 2. Out of all the interestingness measures, is there a definitive best one to use? Why or why not? (review)

---

No, there is **not** one definitive best interestingness measure.

That is because different measures capture different aspects of what makes a pattern “interesting.” A pattern may be interesting because it is:

- common
- reliable
- surprising
- statistically significant
- useful for prediction
- non-redundant

Different measures emphasize different things:

- **Support** emphasizes frequency
- **Confidence** emphasizes rule reliability
- **Lift** emphasizes departure from independence 
- $x^2$ emphasizes statistical association
- **Closed/max patterns** emphasize compactness and non-redundancy
- **Mutual information** emphasizes shared information

The best choice depends on the goal.

For example:

- If you want patterns that occur often, support matters.
- If you want good recommendation rules, confidence and lift may matter more.
- If you want statistically meaningful relationships, $x^2$ may be more useful.
- If you want a compact summary of patterns, closed or maximal itemsets may be better.

So there is no universally best measure because “interestingness” is context-dependent.

### 3. What does mutual information and pointwise mutual information specifically measure? (review)

---

**Mutual information (MI)** measures how much knowing one variable reduces uncertainty about another. In other words, it measures how much information two variables share.

If two variables are independent, knowing one tells you nothing about the other, so the mutual information is low or zero. If they are strongly related, the mutual information is higher.

So mutual information measures the **overall dependence** between two variables.

**Pointwise mutual information (PMI)** is more specific. It measures how strongly a **particular pair of outcomes** is associated compared to what would be expected if they were independent.

For two events X and Y, PMI is:
$
PMI(X, Y) = \log \frac {P(X, Y)}{P(X)P(Y)}
$

Interpretation:
- **PMI > 0**: `X` and `Y` co-occur more than expected
- **PMI = 0**: they are independent
- **PMI < 0**: they co-occur less than expected

The difference is:

- **Mutual information** summarizes the overall relationship across all outcomes
- **Pointwise mutual information** focuses on one specific event pair

A simple way to think about it:

- **MI** = “How related are these variables overall?”
- **PMI** = “How surprising is this specific co-occurrence?”

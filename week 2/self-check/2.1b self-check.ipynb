### 1. Simply counting all possible frequent itemsets is the simplest way to find them, but what is the problem with this method? (review)

---

The problem is that the number of possible itemsets grows **combinatorially** as the number of items increases. Even with a moderate number of unique items, the total number of possible subsets becomes enormous.

For example, if there are $𝑛$ distinct items, there are $2^{𝑛−1}$ possible non-empty itemsets. That means checking every possible combination quickly becomes too expensive in both **time** and **memory**.

This brute-force method is impractical for large datasets because most candidate itemsets will turn out to be infrequent, yet the algorithm would still waste resources counting them.

### 2. What is the downward closure property and how does it relate to the Apriori algorithm? (review)

---

The **downward closure property** says:

- If an itemset is **frequent**, then **all of its subsets must also be frequent**.
- Equivalently, if an itemset is **infrequent**, then **all of its supersets must also be infrequent**.

This property is the key idea behind the **Apriori algorithm**. Apriori uses it to avoid checking itemsets that cannot possibly be frequent.

For example, if 
`
{milk, bread} 
`
is not frequent, then 
`
{milk, bread, eggs}
`
cannot be frequent either. So Apriori prunes that larger candidate immediately instead of wasting time counting it.

### 3. How does the Apriori algorithm solve the problem with finding frequent itemsets in large sets of data? (review)

---

Apriori solves the problem by reducing the number of itemsets it needs to consider.

Instead of counting every possible itemset, it works **level by level**:

- First, it finds all frequent **1-itemsets**
- Then it uses those to generate candidate **2-itemsets**
- Then frequent **2-itemsets** are used to generate candidate **3-itemsets**
- And so on

At each step, Apriori prunes candidates using the downward closure property. If any subset of a candidate is not frequent, that candidate is discarded before counting support.

This makes the search much more efficient because the algorithm focuses only on combinations that still have a realistic chance of being frequent.

### 4. Draw or write out a simple high-level explanation of how the Apriori algorithm works in your mind. Later, see how this compares to the implementation in the following assignment.

---

**How I think of Apriori**:

1. Start with single items.
2. Count how often each item appears.
3. Keep only the items that meet minimum support.
4. Combine the surviving items to make larger candidates.
5. Before counting them, throw out any candidate whose smaller subsets were not frequent.
6. Count support for the remaining candidates.
7. Keep only the frequent ones.
8. Repeat until no new frequent itemsets can be found.

A simple mental model is:

- **Find small frequent groups first**
- **Use them to build bigger groups**
- **Stop exploring branches that already failed**

You can also picture it like a filter:

- Level 1: frequent single items
- Level 2: build pairs from those items
- Level 3: build triples only from frequent pairs
- Continue until nothing survives

A compact version you could write is:

Apriori:
1. Find all frequent 1-itemsets
2. Use them to generate candidate 2-itemsets
3. Count supports and keep only frequent 2-itemsets
4. Use frequent 2-itemsets to generate candidate 3-itemsets
5. Repeat until no more frequent itemsets are found

The main idea is that Apriori does **not** search everything. It uses earlier results to eliminate impossible candidates early.

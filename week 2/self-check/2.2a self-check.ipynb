### 1. Can you define max and closed patterns and give examples of each? Are all max patterns, closed patterns, or are all closed patterns, max patterns? (review)

---

A **max pattern** (or **maximal frequent pattern**) is a frequent itemset that has no frequent superset.

A **closed pattern** is a frequent itemset for which **no superset has exactly the same support**.

Example:

Suppose these itemsets are frequent:

- `{A}` support = 5
- `{B}` support = 4
- `{A, B}` support = 4
- `{A, B, C}` support = 4

Then:

- `{A, B, C}` is **maximal** if there is no larger frequent itemset containing it.
- `{A, B}` is **not maximal** because it has a frequent superset: `{A, B, C}`.

Now look at closedness:

- `{A, B}` has support 4
- `{A, B, C}` also has support 4

Because a superset of `{A, B}` has the **same support**, `{A, B}` is **not closed**.

- `{A, B, C}` is **closed** because no larger frequent superset has the same support.

Relationship:

- **All maximal patterns are closed**
- **Not all closed patterns are maximal**

Why?
If a pattern is maximal, it has no frequent superset at all, so it definitely cannot have a frequent superset with the same support. But a closed pattern can still have a larger frequent superset, as long as that larger superset has **lower** support.

### 2. Can max and closed patterns be used to measure the “interestingness” of a pattern? Can association rules? (review)

---

Yes, but in different ways.

**Max** and **closed** patterns help reduce redundancy and highlight patterns that carry more useful information.

- **Maximal patterns** are interesting because they summarize the largest frequent combinations without listing every smaller subset.
- **Closed patterns** are interesting because they preserve support information while removing redundant frequent itemsets.

So they do not directly measure interestingness the way a score does, but they help identify patterns that are often **more informative** and **less repetitive**.

**Association rules** can also be used to measure interestingness, and in fact they are often more directly tied to usefulness because they evaluate relationships of the form:

`X → Y`

These rules can be judged using measures such as:

- support
- confidence
- lift
- correlation-related measures

So yes, association rules can be used to assess interestingness, especially when the goal is to find meaningful or actionable relationships.

### 3. Why can association rules be used for recommendations and classification? What feature(s) help them measure the usefulness of a pattern? (review)

---

Association rules can be used for **recommendations** because they capture co-occurrence relationships.

Example:
`
{bread, peanut butter} → {jelly}
`

This can support a recommendation like: “People who bought bread and peanut butter also often bought jelly.”

They can be used for **classification** when the rule predicts a class label.

Example:
`
{high_income, owns_home} → {responds_yes}
`

This works like a simple rule-based classifier: if the left-hand side is true, predict the class on the right-hand side.

The main features that help measure usefulness are:

- **Support**: how often the rule appears overall
- **Confidence**: how often the right-hand side is true when the left-hand side is true
- **Lift**: whether the relationship is stronger than random chance

These help determine whether a rule is common enough, reliable enough, and meaningful enough to be useful.

### 4. Why are support and confidence alone not sufficient to measure the true interestingness of patterns? (review)

---

Support and confidence are helpful, but they can be misleading.

**Support** only tells us how common a pattern is. A pattern can be frequent simply because the items involved are individually very common, not because there is any meaningful relationship between them.

**Confidence** can also be misleading because it depends heavily on how common the consequence is.

Example:

If 90% of customers buy milk, then the rule
`
{bread} → {milk}
`
may have high confidence even if buying bread has little to do with buying milk. Milk is just common overall.

So support and confidence do not tell us whether the relationship is actually **surprising, informative**, or **stronger than chance**.

That is why other measures, especially **lift**, are used to judge whether a rule reflects real association rather than just high base rates.

### 5. What is correlation and how is lift used to measure correlation? (review)

---

**Correlation** in this context means whether two items occur together **more often or less often than expected by chance**.

If two items appear together more often than expected from their individual frequencies, they have a **positive association**.
If they appear together less often than expected, they have a **negative association**.

Lift is a measure that compares:

- the observed co-occurrence of `X` and `Y`
to
- the co-occurrence we would expect if `X` and `Y` were independent

Formula:
$
Lift(X→Y) = \frac{support(X\cup Y)}{support(X)support(Y)}
$

Interpretation:

- **Lift** = 1: `X` and `Y` are independent
- **Lift > 1**: positive correlation; they occur together more than expected
- **Lift < 1**: negative correlation; they occur together less than expected

Example:

- support(A) = 0.5
- support(B) = 0.4
- support(A and B) = 0.3

Then:
$
Lift = \frac{0.3}{0.5 \times 0.4} = \frac{0.3}{0.2} = 1.5
$

Since lift is greater than 1, A and B occur together more often than expected by chance, so they are positively correlated.

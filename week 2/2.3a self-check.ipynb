### 1. Why can't we just use the number of items shared between two different itemsets as a measure of their similarity? (review)

---

We cannot rely only on the **number of shared items** because that ignores the overall size of the two itemsets.

For example:

- Set A = {milk, bread}
- Set B = {milk, bread, eggs, cereal, juice}

These two sets share 2 items. Now compare that with:

- Set C = {milk, bread}
- Set D = {milk, bread}

These also share 2 items.

If we only count shared items, both pairs look equally similar. But clearly C and D are identical, while A and B are only partially similar. The raw intersection count misses that difference.

So the problem is that shared-item count is **not normalized**. Larger sets naturally have more opportunities to overlap, which can make them seem more similar than they really are. A good similarity measure should consider both:

- how much the sets overlap
- how large the sets are overall

That is why measures like **Jaccard similarity** are more useful.

### 2. Can you define and give examples of an intersection and union? (review)

---

Yes.

The **intersection** of two sets is the set of items they have **in common**.

Example:

- A = {bread, milk, eggs}
- B = {milk, eggs, juice}

Then:

- $ A∩B = {{milk,eggs}} $

So the intersection contains only the items that appear in both sets.

The **union** of two sets is the set of **all distinct items** that appear in either set.

Using the same example:

- A = {bread, milk, eggs}
- B = {milk, eggs, juice}

Then:

- $ A∪B = {{bread, milk, eggs, juice}} $

So the union includes everything from both sets, without duplicates.

A simple way to remember it:

- **intersection** = overlap
- **union** = everything combined

### 3. Why are intersections on the numerator and unions in the denominator of the calculation for Jaccard Similarity? (review)

---

Jaccard Similarity is defined as:
$ 
J (A, B) = \frac{|A\cap B|}{|A\cup B|}
$

The **intersection** is in the numerator because it represents the amount of **shared content** between the two sets. This is the part that makes the sets similar.

The **union** is in the denominator because it represents the **total distinct content** across both sets. This gives the full context for judging how large the overlap really is.

So Jaccard is basically asking:

**Out of everything that appears in either set, what fraction is shared by both?**

This makes it a normalized measure. It prevents large sets from looking artificially similar just because they happen to share several items.

For example:

- A = {1, 2}
- B = {1, 2, 3, 4, 5}

Intersection size = 2
Union size = 5
$
J(A, B) = \frac{2}{5} = 0.4
$
That gives a more accurate picture than just saying they share 2 items.


### 4. How does using Jaccard Similarity and Distance compare to using support? How can both of these be used together for complex data mining tasks? (review)

---

**Support** and **Jaccard similarity** measure different things.

**Support** measures how often an itemset appears in the entire dataset.

Example:

If `{milk, bread}` appears in 30 out of 100 transactions, its support is 30 or 30%.

So support is about **frequency in the database**.

**Jaccard similarity** measures how similar **two specific itemsets or two records** are, based on overlap.

So Jaccard is about **pairwise similarity**, not overall frequency.

**Jaccard distance** is just the opposite idea:

$
Jaccard Distance = 1 - Jaccard Similarity 
$

It measures how dissimilar two sets are.

Comparison:

- **Support**: “How common is this pattern in the dataset?”
- **Jaccard similarity**: “How similar are these two transactions, users, or itemsets?”
- **Jaccard distance**: “How different are they?”

They can be used together in more complex tasks because they answer different questions.

For example:

- Use **support** to find frequent itemsets in market basket data
- Use **Jaccard similarity** to compare customers based on what they buy
- Then group similar customers or recommend items based on similar baskets

Another example:

- Support can identify important recurring patterns
- Jaccard can compare those patterns or compare records containing them

So together:

- **support** helps identify patterns that matter globally
- **Jaccard** helps compare specific sets locally

This combination is useful in clustering, recommendation systems, nearest-neighbor methods, and exploratory pattern mining.

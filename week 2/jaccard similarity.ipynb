### What is "similarity"?

- Distance measures the opposite: how much they are dissimilar

### Intersection and Union 

- Intersection (**$A \cap B$**): largest common subset of A and B
- Union (**$A \cup B$**): smallest common superset of A and B

{:beer:, :baby_bottle:, :watermelon:} $\cap$ {:baby_bottle:, :beer:, :lemon:} = {:beer:, :baby_bottle:}

{:beer:, :baby_bottle:, :watermelon:} $\cup$ {:baby_bottle:, :beer:, :lemon:} = {:beer:, :baby_bottle:, :watermelon:, :lemon:}

### The Jaccard Similarity

- A simple (but powerful) measure of similarity of two sets
    - also known as Jaccard coefficient and Jaccard index 



Jaccard distance: $1 - J(A, B)$

### Properties of Jaccard Similaity 

- $J(A, B) = J(B, A)$
- $0 \leq J(A, B) \leq 1$
- $J(A, B) = 0$ if two sets share no items
- $J(A, B) = 1$ if two sets are identical
